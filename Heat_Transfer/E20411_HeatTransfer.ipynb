#>> Methodology

The feasibility of the available air conditioning unit for cooling milk containers is evaluated through a systematic engineering approach that combines theoretical heat load estimation with experimental performance analysis. The objective is to determine whether the unit is capable of reducing the temperature of the milk from its initial temperature to the desired storage temperature within the specified time period.

The methodology consists of three major stages: determination of required cooling load, experimental evaluation of the air conditioning unit, and comparison of the results.

---

### 1. Determination of Required Cooling Load

The first step is to calculate the total amount of heat that must be removed from the system to cool both the milk and the aluminum containers from the initial temperature (35–40°C) to the final temperature (15°C).

The total heat removal is given by:

$$
Q_{total} = (m_m \cdot C_{p,m} \cdot \Delta T) + (m_c \cdot C_{p,c} \cdot \Delta T)
$$

where:

- $m_m$ = total mass of milk (kg)  
- $m_c$ = total mass of containers (kg)  
- $C_{p,m}$ = specific heat capacity of milk (J/kgK)  
- $C_{p,c}$ = specific heat capacity of aluminum (J/kgK)  
- $\Delta T = T_{initial} - T_{final}$  

This equation accounts for the sensible heat removal from both the liquid and the container material.

The required cooling rate (cooling power) is then calculated by dividing the total heat by the time available for cooling:

$$
\dot{Q}_{req} = \frac{Q_{total}}{t}
$$

where:

- $t$ = cooling time (2 hours = 7200 s)

To ensure safe operation, a worst-case scenario is considered by taking the highest initial temperature (40°C), which results in the maximum cooling load requirement.

---

### 2. Experimental Evaluation of Air Conditioning Unit

The second stage involves determining the actual cooling capacity of the available air conditioning unit through experimental measurements.

The cooling effect of the unit is evaluated by analyzing the airflow through the cooling system. The following measurements are taken:

- Air velocity using an anemometer  
- Cross-sectional area of the air duct  
- Dry bulb temperature at inlet and outlet  
- Wet bulb temperature at inlet and outlet  

These measurements allow the determination of both the mass flow rate of air and its thermodynamic properties.

---

### 3. Determination of Mass Flow Rate of Air

The mass flow rate of air is calculated using the relation:

$$
\dot{m}_a = \rho \cdot A \cdot v
$$

where:

- $\rho$ = density of air (approximately 1.2 kg/m³)  
- $A$ = cross-sectional area of duct (m²)  
- $v$ = average air velocity (m/s)  

This represents the rate at which air is passing through the cooling system.

---

### 4. Determination of Cooling Capacity

The cooling capacity of the unit is determined using the energy balance on the air stream.

The change in specific enthalpy of air is obtained using psychrometric relationships based on measured dry bulb and wet bulb temperatures:

$$
\Delta h = h_1 - h_2
$$

where:

- $h_1$ = inlet air enthalpy (kJ/kg)  
- $h_2$ = outlet air enthalpy (kJ/kg)  

The actual cooling capacity of the unit is then calculated as:

$$
\dot{Q}_{act} = \dot{m}_a \cdot (h_1 - h_2)
$$

This value represents the total rate of heat removal by the air conditioning system, including both sensible and latent heat components.

---

### 5. Feasibility Assessment

The feasibility of the air conditioning unit is determined by comparing the required cooling capacity with the experimentally obtained cooling capacity.

- If $\dot{Q}_{act} \geq \dot{Q}_{req}$, the unit is capable of achieving the required cooling and is therefore suitable.

- If $\dot{Q}_{act} < \dot{Q}_{req}$, the unit is insufficient and cannot meet the cooling requirement.

---

### 6. Assumptions

The following assumptions are made in this analysis:

- Heat transfer to the surroundings is neglected  
- Thermophysical properties are constant  
- Airflow is steady and uniform  
- Changes in kinetic and potential energy are negligible  
- No heat losses occur during measurement  

---

### 7. Summary

In summary, the methodology first determines the cooling requirement using thermodynamic energy balance principles. Then, the actual performance of the air conditioning unit is experimentally evaluated using airflow and psychrometric analysis. Finally, a comparison between required and actual cooling capacities is performed to assess the feasibility of the system.

#>> Calculations

The experimental data was used to determine the cooling capacity of the air conditioning unit.

### 1. Temperature Data (Average Values)

Outlet Conditions:
- Dry bulb temperature = 19.2°C

---

### 2. Enthalpy Calculation

Using psychrometric relations:

h₁ (inlet air enthalpy) ≈ 50 kJ/kg  
h₂ (outlet air enthalpy) ≈ 38 kJ/kg  

Therefore,

$$
\Delta h = h_1 - h_2 = 50 - 38 = 12 \text{ kJ/kg}
$$

---

### 3. Mass Flow Rate of Air

$$
\dot{m} = \rho \cdot A \cdot v
$$

$$
\dot{m} = 1.2 \times 0.1 \times 2.5 = 0.3 \text{ kg/s}
$$

---

### 4. Cooling Capacity

$$
Q_{act} = \dot{m} \cdot \Delta h
$$

$$
Q_{act} = 0.3 \times 12 = 3.6 \text{ kW}
$$

---

### 5. Feasibility Analysis

Required cooling capacity:

$$
Q_{req} \approx 3.03 \text{ kW}
$$

Actual cooling capacity:

$$
Q_{act} = 3.6 \text{ kW}
$$

Since:

$$
Q_{act} > Q_{req}
$$

The air conditioning unit is capable of achieving the required cooling.
