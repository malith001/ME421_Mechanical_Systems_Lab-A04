# Cooling Load Calculation

This section determines the required cooling capacity to chill 200 kg of milk and 40 aluminum containers from an initial temperature to a target storage temperature within 2 hours.

---

## 1.1 Input Parameters

| Parameter | Symbol | Value | Unit |
| --- | --- | --- | --- |
| Number of Containers | n | 40 | — |
| Total Mass of Milk | m_m | 200 | kg |
| Total Mass of Containers | m_c | 80 | kg |
| Specific Heat Capacity of Milk | C_p,m | 4000 | J/kg·K |
| Specific Heat Capacity of Aluminum | C_p,c | 921.1 | J/kg·K |
| Temperature Reduction (Normal) | ΔT | 22.5 | K |
| Cooling Time | t | 7200 | s |

---

## 1.2 Total Energy Calculation

The total heat energy that must be removed is the sum of energy extracted from both the milk and the containers.

**Formula:**

Q_total = (m_m × C_p,m × ΔT) + (m_c × C_p,c × ΔT)

**Step-by-Step Calculation:**

| Component | Formula | Calculation | Result (J) |
| --- | --- | --- | --- |
| Milk | m_m × C_p,m × ΔT | 200 × 4000 × 22.5 | 18,000,000 |
| Containers | m_c × C_p,c × ΔT | 80 × 921.1 × 22.5 | 1,657,980 |
| Total | Q_total | Sum of above | 19,657,980 |

Q_total = 19.66 MJ

---

## 1.3 Required Cooling Power

The cooling power required is the total energy divided by the time available.

**Formula:**

Q_dot_req = Q_total / t

| Parameter | Value |
| --- | --- |
| Total Energy (Q_total) | 19,657,980 J |
| Time (t) | 7200 s |
| Required Power (Q_dot_req) | 2730.28 W |

Q_dot_req = 2.73 kW

Note: This value represents the minimum sensible cooling capacity required, excluding external heat gains from the environment.

---

## 1.4 Worst-Case Scenario Analysis

For a higher initial temperature of 40°C, the temperature reduction increases to ΔT = 25 K.

**Parameters for Worst-Case:**

| Parameter | Value |
| --- | --- |
| Initial Temperature | 40°C |
| Final Temperature | 15°C |
| Temperature Change (ΔT) | 25 K |

**Total Energy Calculation:**

Q_max = (200 × 4000 × 25) + (80 × 921.1 × 25)

| Component | Calculation | Result (J) |
| --- | --- | --- |
| Milk | 200 × 4000 × 25 | 20,000,000 |
| Containers | 80 × 921.1 × 25 | 1,842,200 |
| Total | Q_max | 21,842,200 |

Q_max = 21.84 MJ

**Required Maximum Cooling Power:**

Q_dot_max = 21,842,200 / 7200 = 3033.64 W

Q_dot_max = 3.03 kW

---

## 1.5 Summary of Results

| Scenario | Temperature Change (ΔT) | Total Energy (Q) | Required Power (Q_dot) |
| --- | --- | --- | --- |
| Normal Operation | 22.5 K | 19.66 MJ | 2.73 kW |
| Worst-Case (40°C) | 25.0 K | 21.84 MJ | 3.03 kW |

---


### 2.1 Test Apparatus and Sensors

Airflow properties are measured using a standard duct-traverse technique to fully characterize the air stream's thermo-physical state.

- **Duct Dimensions:** Cross-section area $A = w \times h$.
- **Flow Speed:** Point-wise velocities ($v$) obtained with multi-point anemometer grid.
- **Air Conditions:** Dry-bulb ($T_{db}$) and wet-bulb ($T_{wb}$) temperatures logged at **inlet (point 1)** and **outlet (point 2)**.



### 2.2 Calculation Procedure

Cooling output is computed from air mass flow and the enthalpy drop across the unit.

**Step 1: Air Mass Flow Rate ($\dot{m}_a$)**  
$$\dot{m}_a = \rho \cdot A \cdot \bar{v}$$  
*(Here, $\rho \approx 1.2 \text{ kg/m}^3$ is air density; $\bar{v}$ is average velocity.)*

**Step 2: Actual Cooling Rate ($\dot{Q}_{act}$)**  
Total cooling (sensible + latent) follows the energy balance:  
$$\dot{Q}_{act} = \dot{m}_a (h_1 - h_2) = \dot{m}_a \Delta h$$  
*Enthalpies $h_1$ and $h_2$ come from psychrometric charts using measured $T_{db}$ and $T_{wb}$.*

## 4. Thermodynamic Analysis Framework

This section describes the analytical procedure used to convert raw sensor data into the system’s **Actual Cooling Capacity** ($\dot{Q}_{act}$).

### 4.1 Air Mass Flow Rate ($\dot{m}_a$)

The mass flow of air is the primary driver of heat transfer. It is obtained from the duct geometry and inlet velocity using the mass‑flow form of the continuity equation:
$$
\dot{m}_a = \rho \cdot A \cdot V_{in}
$$
*(where $\rho \approx 1.2 \text{ kg/m}^3$ is air density, $A$ is the duct cross‑section area, and $V_{in}$ is the measured inlet air velocity).*

### 4.2 Psychrometric Enthalpy Model

To capture both **sensible** (temperature) and **latent** (moisture) components of cooling, the specific enthalpy $h$ is derived from the Magnus–Tetens saturation‑vapor‑pressure model.

#### Saturation and Partial Vapor Pressure
First, the saturation vapor pressure at the wet‑bulb temperature $T_{wb}$ is computed:
$$
P_{swb} = 0.61078 \cdot \exp\left(\frac{17.27 \cdot T_{wb}}{T_{wb} + 237.3}\right)
$$
Then the actual water‑vapor partial pressure $P_v$ is estimated as:
$$
P_v = P_{swb} - 0.000662 \cdot P_{atm} \cdot (T_{db} - T_{wb})
$$

#### Humidity Ratio and Specific Enthalpy
The humidity ratio $\omega$ (kg vapor per kg dry air) is:
$$
\omega = 0.62194 \cdot \frac{P_v}{P_{atm} - P_v}
$$
From this, the total specific enthalpy of moist air is:
$$
h = (c_{p,a} \cdot T_{db}) + \omega \cdot (h_{fg} + c_{p,v} \cdot T_{db})
$$
> **Constants:**  
> $c_{p,a} = 1.006 \text{ kJ/kg·K}$,  
> $h_{fg} = 2501 \text{ kJ/kg}$,  
> $c_{p,v} = 1.86 \text{ kJ/kg·K}$,  
> $P_{atm} = 101.325 \text{ kPa}$.

### 4.3 System Cooling Capacity ($\dot{Q}_{act}$)

The **Actual Cooling Capacity** is obtained by multiplying the air mass flow rate by the enthalpy difference across the evaporator:
$$
\dot{Q}_{act} = \dot{m}_a \cdot (h_{inlet} - h_{outlet})
$$
*(where $h_{inlet}$ and $h_{outlet}$ are the specific enthalpies at the evaporator inlet and outlet, respectively, computed from the measured $T_{db}$ and $T_{wb}$ pairs).*

### 4.4 Performance Index (Efficiency Ratio)

The effectiveness of each fan‑speed setting is summarized by a **Performance Ratio** $\eta$, defined as the ratio of delivered cooling to the required design load:
$$
\eta = \frac{\dot{Q}_{act}}{\dot{Q}_{req}}
$$
where $\dot{Q}_{req} = 3.03 \text{ kW}$ is the target thermal load. A value of $\eta \geq 1$ indicates that the system meets the design requirement at that speed.


```python
import numpy as np
import pandas as pd

def calculate_enthalpy(Tdb, Twb):
    """
    Calculates specific enthalpy (h) in kJ/kg based on Dry Bulb and Wet Bulb temperatures.
    Uses the formulas established in Section 5.
    """
    Patm = 101.325  # kPa

    # I. Saturation Vapor Pressure at Wet Bulb (Pswb)
    Pswb = 0.61078 * np.exp((17.27 * Twb) / (Twb + 237.3))

    # II. Actual Vapor Pressure (Pv)
    Pv = Pswb - 0.000662 * Patm * (Tdb - Twb)

    # III. Humidity Ratio (w)
    w = 0.62194 * (Pv / (Patm - Pv))

    # IV. Total Specific Enthalpy (h)
    # cp_a = 1.006, hfg = 2501, cp_v = 1.86
    h = (1.006 * Tdb) + w * (2501 + 1.86 * Tdb)

    return h

# Experimental Data
data = {
    'Speed': range(1, 11),
    'T3_db': [29.1, 29.1, 29.5, 29.6, 29.3, 29.3, 29.3, 29.1, 28.9, 29.3],
    'T4_wb': [21.2, 21.1, 21.5, 21.8, 21.2, 21.1, 21.1, 21.0, 20.9, 21.1],
    'T5_db': [26.4, 26.7, 24.3, 20.4, 20.8, 21.4, 21.5, 21.2, 21.3, 19.9],
    'T6_wb': [19.9, 20.0, 19.7, 16.6, 16.2, 18.0, 18.2, 17.6, 17.4, 16.5],
    'Vin':   [0.0, 0.0, 0.6, 2.8, 5.6, 8.8, 9.3, 9.0, 9.8, 9.5]
}

df = pd.DataFrame(data)

# Constants
rho = 1.2  # kg/m3
A = 0.05   # EDIT THIS: Enter your actual Duct Area in m^2

# Process Calculations
df['h_in'] = df.apply(lambda row: calculate_enthalpy(row['T3_db'], row['T4_wb']), axis=1)
df['h_out'] = df.apply(lambda row: calculate_enthalpy(row['T5_db'], row['T6_wb']), axis=1)
df['m_dot'] = rho * A * df['Vin']
df['Q_act_kW'] = df['m_dot'] * (df['h_in'] - df['h_out'])

# Comparison against Design Load (3.03 kW)
df['Requirement_Met'] = df['Q_act_kW'] >= 3.03

# Results
print("--- Evaporator Performance Analysis ---")
display(df[['Speed', 'Vin', 'h_in', 'h_out', 'Q_act_kW', 'Requirement_Met']])
```

## 5. System Scaling for 80‑Container Operation

When the number of containers doubles, the total heat energy to be removed also doubles within the same time window.

**Total heat energy to remove ($Q_{new}$):**  
$$
Q_{new} = 2 \cdot Q_{max} = 2 \cdot 21{,}842{,}200 \text{ J}

$$

\dot{Q}_{req,80} = \frac{Q_{new}}{7200} \approx \mathbf{6.06 \text{ kW}}
