**Known Data:**

---


1. Number of containers = 40

Per container:

*   Milk mass = 5 kg
*   Container mass = 2 kg (Aluminum)

---



2. Specific heats:


*   Milk: $C_{p,\text{milk}}$ = 4000J/kg⋅K
*   Aluminum: $C_{p,\text{Al}}$ = 921.1J/kg⋅K

---



3. Target cooling:

*   $T_{initial}$ = 35–40°C
*   $T_{final}$ = 15°C

---



4. Worst case used for design:

*   $T_{initial}$ = 40°C
*   ΔT= 40−15 =25K

---



5. Heat Removal per Container:

Milk cooling load,

<center>
$Q_{milk}$ = m$C_{p}$ ​ΔT

$Q_{milk}$ = 5×4000×25

$Q_{milk}$ ​= 500,000J
</center>


Container cooling load,
<center>
$Q_{Al}$ = 2×921.1×25

$Q_{Al}$ = 46,055J
</center>

Total per container,
<center>
$Q_{container, total}$ = 500,000J+ 46,055J =546,055J
</center>

---



Total Heat Removal (40 containers),
<center>
$Q_{total}$ = 40×546,055

$Q_{total}$ = 21,842,200J

$Q_{total}$ ≈ 21.84MJ
</center>
Required Cooling Rate,

*   Cooling time = 2 hours = 7200 s

<center>
$\dot{Q}$ ​= Q/t

$\dot{Q} = $$\frac{21.84 \times 10^{6}}{7200}$

$\dot{Q}$ ​≈ 3033W

$\dot{Q}$ ≈ 3.03kW
</center>


---


Minimum refrigeration capacity required ≈ 3.1 kW.

---



Applying Safety Factors,

Include:



* Milk loading disturbance
* Fan & lighting loads
* Air infiltration
* Heat gains from walls
* Door opening losses





Typical safety margin:
<center>
1.2–1.5
</center>

Assume 1.4:
<center>

$Q_{design​}$ = 3.03×1.4

$Q_{design​}$ ≈ 4.24kW

</center>

Design cooling capacity required ≈ 4.3 kW

---



**Comparing with AC Unit Capacity**

Obtain AC nameplate data:

*   Cooling capacity (kW or BTU/hr)

Conversion:
*   1kW = 3412BTU/hr

Feasibility Criteria:

| Condition | Interpretation |
|----------|----------|
| AC < 3 kW       | Cannot meet 2-hour cooling        |
| AC ≈ 3–4 kW     | Marginal       |
| AC > 4.3 kW   | Suitable       |      



---
Evaluating Temperature Maintenance Load

After reaching 15 °C:

Only heat gains must be offset:

Sources:

*   Transmission through walls
*   Air leakage
*   Door opening
*   Warm container addition (if any)

Method:

<center>
$Q_{maintenance​​}$ =UA( $T_{ambient​}$-$T_{room​​}$)
</center>

Where:
*   ΔT = Temperature difference
*   A = Surface area
*   U = Overall heat transfer coefficient

If:

$Q_{maintenance​​}$ < $AC_{capacity}$

Temperature can be maintained.
