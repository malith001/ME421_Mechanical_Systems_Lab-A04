## Part 1: Propose a method to find out the feasibility of the available unit.  

\subsection*{Proposed method to find out the feasibility of the available unit}

The feasibility of the available Air Conditioning unit can be checked by following the steps below.

\begin{enumerate}
    \item \textbf{Collect the required data} \\
    First, the given data should be collected. These include the number of milk containers received per day, mass of milk in one container, mass of the empty aluminum container, specific heat capacities of milk and aluminum, initial temperature of the milk, required final temperature, and the time allowed for cooling.

    \item \textbf{Consider the worst operating condition} \\
    For the analysis, the highest initial milk temperature should be taken. In this case, it is $40^\circ\text{C}$. Since the milk has to be cooled to $15^\circ\text{C}$, the unit should be checked for this worst condition.

    \item \textbf{Calculate the heat removed from the milk} \\
    The amount of heat that must be removed from the milk should be calculated by using the sensible heat equation. For this, the total mass of milk in all containers should be taken, and then the heat removed to reduce the temperature from $40^\circ\text{C}$ to $15^\circ\text{C}$ should be found.

    \item \textbf{Calculate the heat removed from the aluminum containers} \\
    In the same way, the heat that must be removed from the aluminum containers should also be calculated. The total mass of all containers should be considered here.

    \item \textbf{Find the total cooling load} \\
    After that, the heat removed from the milk and the heat removed from the containers should be added together. This gives the total cooling load that the unit must handle.

    \item \textbf{Determine the required cooling capacity} \\
    Since the milk has to be cooled within 2 hours, the total cooling load should be divided by the cooling time. This gives the minimum cooling capacity required from the unit.

    \item \textbf{Apply a safety factor} \\
    In actual operation, extra heat can enter the system due to wall heat transfer, air leakage, door opening, lighting, and other disturbances. Therefore, a safety factor should be added to the calculated cooling capacity to get a safer design value.

    \item \textbf{Find the actual capacity of the available AC unit} \\
    Then the actual cooling capacity of the available AC unit in the Heat Engine Laboratory should be found from the nameplate, manufacturer data, or laboratory manual.

    \item \textbf{Compare the required capacity with the available capacity} \\
    Next, the required cooling capacity should be compared with the available AC unit capacity. If the available capacity is higher than the required value, the unit can be considered suitable. If it is lower, the unit is not suitable for the required task.

    \item \textbf{Check the ability to maintain $15^\circ\text{C}$} \\
    After cooling the milk to $15^\circ\text{C}$, the unit must keep it at that temperature until the next day at 1400 h. Therefore, the maintenance load should also be checked by considering the heat entering from the surroundings.

    \item \textbf{Carry out an experimental test} \\
    An experiment should be done using the laboratory cooling unit. Water-filled containers can be used instead of milk to represent the same thermal load. During the test, measurements such as initial temperature, room temperature, load temperature at regular time intervals, time taken to reach $15^\circ\text{C}$, and power consumption should be recorded.

    \item \textbf{Verify the feasibility experimentally} \\
    From the experimental results, it should be checked whether the unit can cool the load to $15^\circ\text{C}$ within 2 hours and maintain that temperature steadily without overloading.

    \item \textbf{Extend the method for 80 containers per day} \\
    The same procedure should then be repeated for 80 containers per day. If the required cooling load becomes higher than the available capacity of the unit, then modifications will be needed.

    \item \textbf{Propose developments if required} \\
    If the available unit is not enough, improvements such as using a higher-capacity cooling unit, improving insulation, increasing airflow circulation, reducing door opening, or pre-cooling the milk can be suggested.

    \item \textbf{Draw the final conclusion} \\
    Finally, a conclusion should be given based on both the theoretical calculations and the experimental results to decide whether the available unit is suitable for cooling and storing the milk containers.
\end{enumerate}
