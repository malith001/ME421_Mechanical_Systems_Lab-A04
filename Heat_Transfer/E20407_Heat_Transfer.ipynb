# Problem Statement

## The following data have been provided:

- Weight of an empty container: **2 kg**

## 1. Experimental Setup Overview

The experimental apparatus consists of two primary subsystems that work in conjunction to achieve cooling:

### 1.1 Refrigeration Circuit
The system operates on a standard vapor compression refrigeration cycle using R22 refrigerant. The main components include:

| Component | Description | Function |
|-----------|-------------|----------|
| Compressor | Reciprocating type | Raises refrigerant pressure and temperature |
| Condenser | Air-cooled with fan | Rejects heat to ambient |
| Expansion Valve | Thermostatic | Reduces refrigerant pressure |
| Evaporator | Located in Perspex tunnel | Absorbes heat from circulating air |

### 1.2 Air Circulation System
The air handling components include:
- **Blower:** Circulates air through the system
- **Supply Duct:** 110 mm diameter, delivers cooled air to the conditioned space
- **Return Duct:** Returns warm air to the evaporator
- **Conditioned Space:** Simulates cold room storage

*Note: The electric heating coil installed downstream of the evaporator remained inactive during testing.*

## 2. Theoretical Cooling Load Analysis

### 2.1 Problem Definition
A dairy processing facility requires cooling of milk containers under the following conditions:
- **Initial temperature:** 35°C - 40°C (upper bound used for calculations)
- **Target temperature:** 15°C
- **Cooling duration:** 2 hours
- **Storage duration:** 24 hours at 15°C

### 2.2 Material Properties

| Material | Mass per container | Specific Heat Capacity |
|----------|-------------------|----------------------|
| Milk | 5.0 kg | 4.00 kJ/kg·K |
| Aluminum container | 2.0 kg | 0.921 kJ/kg·K |

### 2.3 Single Container Cooling Requirement

**Temperature reduction:** ΔT = 40°C - 15°C = 25°C

Heat removed from milk:
$$Q_{milk} = m_{milk} \cdot c_{milk} \cdot \Delta T$$
$$Q_{milk} = 5.0 \times 4.00 \times 25 = 500\text{ kJ}$$

Heat removed from container:
$$Q_{Al} = m_{Al} \cdot c_{Al} \cdot \Delta T$$
$$Q_{Al} = 2.0 \times 0.921 \times 25 = 46.05\text{ kJ}$$

Total per container:
$$Q_{total,1} = 500 + 46.05 = 546.05\text{ kJ}$$

### 2.4 Total Cooling Load (40 Containers)

$$Q_{total,40} = 40 \times 546.05 = 21,842\text{ kJ}$$

### 2.5 Required Cooling Power

Cooling time: t = 2 hours = 7,200 seconds

$$\dot{Q}_{required,40} = \frac{21,842 \times 1000}{7,200} = 3,034\text{ W} \approx \boxed{3.03\text{ kW}}$$

This represents the **minimum** cooling capacity required, assuming no heat gains from surroundings and perfect thermal efficiency.

## 3. Experimental Methodology

### 3.1 Measurement Protocol

The cooling system's performance was evaluated using an air-side energy balance approach:

$$\dot{Q}_{cooling} = \dot{m}_{air} (h_{return} - h_{supply})$$

**Parameters measured:**
- Return air dry-bulb temperature (DBT) and wet-bulb temperature (WBT)
- Supply air DBT and WBT
- Air velocity in the duct

### 3.2 Test Procedure

1. **Initial Conditioning:** Cold room was heated to approximately 24°C by allowing controlled heat ingress
2. **Sealing:** Room was sealed to prevent air exchange with surroundings
3. **Data Collection:** Measurements recorded at 1-minute intervals as temperature decreased from 24°C to 20°C

### 3.3 Mass Flow Rate Calculation

Duct cross-sectional area:
$$A = \frac{\pi D^2}{4} = \frac{\pi \times (0.110)^2}{4} = 0.009503\text{ m}^2$$

Mass flow rate:
$$\dot{m}_{air} = \rho \cdot A \cdot V = 1.2 \times 0.009503 \times 9.0 = 0.1026\text{ kg/s}$$

*Air density assumed constant at 1.2 kg/m³ based on average operating temperature.*

### 4.3 Cooling Capacity Calculation

Using the energy balance equation:
$$\dot{Q}(t) = \dot{m}_{air} \cdot \Delta h(t)$$

**Sample calculation at t = 2 minutes:**
$$\dot{Q}_2 = 0.1026 \times (47.0 - 42.5) = 0.1026 \times 4.5 = 0.462\text{ kW}$$

```python
# Cooling Capacity Analysis
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Input parameters
D = 0.110                      # Duct diameter (m)
V_air = 9.0                    # Air velocity (m/s)
rho_air = 1.2                  # Air density (kg/m³)

# Area calculation
A_duct = np.pi * (D/2)**2       # m²

# Mass flow rate
mdot_air = rho_air * A_duct * V_air    # kg/s

# Enthalpy data from psychrometric chart
h_return = np.array([48.0, 47.5, 47.0, 46.5, 46.2, 46.0, 45.7,
                    45.5, 45.2, 45.0, 44.8, 44.6, 44.3])

h_supply = np.array([43.0, 42.8, 42.5, 42.2, 42.0, 41.8, 41.5,
                    41.3, 41.1, 40.9, 40.7, 40.5, 40.2])

# Instantaneous cooling capacity (kW)
Q_dot = mdot_air * (h_return - h_supply)
Q_dot = np.round(Q_dot, 3)

# Create results table
df_results = pd.DataFrame({
    "Time (min)": np.arange(0, 13),
    "Cooling Capacity (kW)": Q_dot
})

print("=" * 50)
print("COOLING CAPACITY ANALYSIS RESULTS")
print("=" * 50)
print(df_results.to_string(index=False))

# Statistical summary
Q_avg = np.mean(Q_dot)
Q_min = np.min(Q_dot)
Q_max = np.max(Q_dot)
Q_std = np.std(Q_dot)

print("\n" + "=" * 50)
print("STATISTICAL SUMMARY")
print("=" * 50)
print(f"Average Cooling Capacity: {Q_avg:.3f} kW")
print(f"Maximum Cooling Capacity: {Q_max:.3f} kW")
print(f"Minimum Cooling Capacity: {Q_min:.3f} kW")
print(f"Standard Deviation: {Q_std:.3f} kW")

# Plot results
plt.figure(figsize=(10, 6))
plt.plot(df_results["Time (min)"], Q_dot, 'bo-', linewidth=2, markersize=8, label='Measured')
plt.axhline(y=Q_avg, color='r', linestyle='--', linewidth=2, label=f'Average = {Q_avg:.3f} kW')
plt.axhline(y=3.03, color='g', linestyle=':', linewidth=2, label='Required for 40 containers (3.03 kW)')
plt.xlabel('Time (minutes)', fontsize=12)
plt.ylabel('Cooling Capacity (kW)', fontsize=12)
plt.title('Instantaneous Cooling Capacity vs Time', fontsize=14)
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```

## 6. System Redesign for 80 Containers

### 6.1 Required Cooling Capacity

For 80 containers, the cooling requirement doubles:
$$\dot{Q}_{required,80} = 2 \times 3.03 = \boxed{6.06\text{ kW}}$$

This represents a **13.7× increase** over the current system's capacity.

### 6.2 Proposed Modifications

| Component | Current | Proposed | Justification |
|-----------|---------|----------|---------------|
| **Compressor** | Small reciprocating | Higher capacity unit | Increases refrigerant mass flow rate |
| **Evaporator** | Plain coil | Finned coil with extended surface | Enhances heat transfer area by 3-5× |
| **Blower** | 9.0 m/s | 12-15 m/s | Higher air velocity improves convective heat transfer |
| **Insulation** | Basic | Enhanced (50-75 mm) | Reduces heat gain from surroundings |

### 6.3 Additional Recommendations

1. **Parallel System Configuration:** Consider installing two cooling units operating in parallel for redundancy and scalability
2. **Air Distribution:** Implement proper air distribution system to eliminate dead zones in the cold room
3. **Temperature Monitoring:** Install multiple temperature sensors for better process control
4. **Energy Recovery:** Consider waste heat recovery from condenser for pre-heating applications

### 6.4 Feasibility Assessment

With the proposed modifications, the system can potentially achieve the required 6.06 kW cooling capacity. A complete system redesign, rather than incremental improvements, is recommended due to the significant capacity gap.

## 7. Conclusion

This experimental study evaluated the cooling capacity of an existing refrigeration unit for dairy milk storage applications. Key findings include:

1. **Current Performance:** The system delivers an average cooling capacity of only 0.442 kW, significantly below the 3.03 kW required for 40 containers.

2. **Capacity Gap:** The measured capacity represents only 14.6% of the requirement, indicating a substantial design deficiency.

3. **Scalability Issue:** For 80 containers (6.06 kW requirement), the current system is inadequate by a factor of 13.7.

4. **Required Modifications:** A comprehensive system redesign is necessary, including compressor upgrade, evaporator enhancement, improved air circulation, and better insulation.

5. **Recommendation:** Due to the magnitude of the required capacity increase, a complete system replacement with parallel units is recommended over incremental modifications.

The analysis demonstrates the importance of matching system capacity to actual load requirements and provides a framework for evaluating and upgrading cooling systems for food storage applications.
