## **Experimental Set-Up**

WhatsApp Image 2026-03-26 at 21.36.12.jpeg



The experimental setup used for this study consists of two main subsystems: the **refrigeration cycle** and the **airflow system**.

**1. Refrigeration System**

The refrigeration side of the apparatus operates on a **vapour compression refrigeration cycle** using R134a as the working fluid. The system includes the following primary components:

* **Compressor:**
  A reciprocating (piston-cylinder type) compressor is used to compress the refrigerant vapor, increasing its pressure and temperature.

* **Condenser:**
  The high-pressure refrigerant vapor flows into the condenser, where heat is rejected to the surrounding environment. A fan is installed to force ambient air over the condenser coils, thereby enhancing the convective heat transfer rate.

* **Expansion Valve:**
  The refrigerant then passes through a thermostatic expansion valve, which is mechanically controlled by a sensing bulb filled with refrigerant. This valve regulates the flow and reduces the pressure of the refrigerant before entering the evaporator.

* **Evaporator:**
  The low-pressure refrigerant enters the evaporator, which is located inside a transparent Perspex tunnel. In this component, the refrigerant absorbs heat from the surrounding air, resulting in cooling.

**2. Airflow System**

The airflow side of the apparatus is designed to circulate air between the conditioned space (cold room) and the evaporator:

* **Blower:**
  A blower is used to circulate air through the system. It delivers cooled air (supply air) into the cold room.

* **Supply Air Path:**
  The cooled air is directed into the cold room through an inlet duct of diameter **110.1 mm**.

* **Conditioned Space:**
  The cold air going through the inlet duct absorbs heat from whatever inside the conditioned space.

* **Return Air Path:**
  Air from the cold room is drawn out through an outlet duct of diameter **110.3 mm** and directed into the Perspex tunnel containing the evaporator.

* **Cooling Process:**
  The return air passes over the evaporator coils, where it is cooled by the refrigerant.

* **Heating Coil (Inactive):**
  After the evaporator, the air flows through an electric heating coil, which is currently non-operational and therefore does not affect the system.

* **Air Recirculation:**
  The cooled air is then directed back to the blower and recirculated into the cold room, completing the cycle.

## **Feasibility of the available unit**

**Problem Overview**<br>
A dairy factory needs to:
* **Cool milk:** $35^\circ\text{C} - 40^\circ\text{C} \rightarrow 15^\circ\text{C}$
* **Timeframe:** Within $2$ hours
* **Storage:** Maintain at $15^\circ\text{C}$ for approximately $24$ hours

**Given Data**
* **Containers per day:** $n = 40$
* **Milk per container:** $m_{\text{milk}} = 5 \text{ kg}$
* **Container mass:** $m_{\text{Al}} = 2 \text{ kg}$ (Aluminum)
* **Specific heat of Aluminum:** $c_{\text{Al}} = 921.1 \text{ J/kg}\cdot\text{K}$
* **Specific heat of Milk:** $c_{\text{milk}} = 4000 \text{ J/kg}\cdot\text{K}$


**Step 1: Cooling Load Calculation**

* **Assumptions**<br>
The following assumptions are made to simplify the analysis:

1.  All milk containers are assumed to be at $40^\circ\text{C}$ (upper bound of given range). Initial temperature: $T_i = 40^\circ\text{C}$
2.  The milk inside each container is assumed to have a uniform temperature (no internal gradients).
3.  The aluminum container is assumed to be at the same temperature as the milk due to its high thermal conductivity.
4.  Heat transfer occurs as: <br>$\text{Milk} \rightarrow \text{Container wall} \rightarrow \text{Air}$.
5.  Well-insulated space: No heat gain from surroundings is considered (minimum required load).

Considering the heat removal from One Container;
* Final temperature: $T_f = 15^\circ\text{C}$
* Temperature drop: $\Delta T = T_i - T_f = 25^\circ\text{C}$

Heat removed from milk
$$Q_{\text{milk},1} = m_{\text{milk}} \cdot c_{\text{milk}} \cdot \Delta T$$
$$Q_{\text{milk},1} = 5 \times 4000 \times 25 = 500,000 \text{ J}$$

Heat removed from aluminum container
$$Q_{\text{Al},1} = m_{\text{Al}} \cdot c_{\text{Al}} \cdot \Delta T$$
$$Q_{\text{Al},1} = 2 \times 921.1 \times 25 = 46,055 \text{ J}$$

Total heat per container
$$Q_{1} = Q_{\text{milk},1} + Q_{\text{Al},1}$$
$$Q_{1} = 500,000 + 46,055 = 546,055 \text{ J}$$

Total Heat Load for 40 Containers
$$Q_{\text{total}} = N \cdot Q_{1}$$
$$Q_{\text{total}} = 40 \times 546,055 = 21,842,200 \text{ J}$$

Required Cooling Rate
Cooling must occur within: $t = 2 \text{ hours} = 7200 \text{ s}$

Cooling Power
$$\dot{Q} = \frac{Q_{\text{total}}}{t}$$
$$\dot{Q} = \frac{21,842,200}{7200} \approx 3034 \text{ W}$$

Final Result
$$\boxed{\dot{Q} \approx 3.03 \text{ kW}}$$

Interpretation
* The system must remove approximately **$3.03 \text{ kW}$** of heat continuously over 2 hours.
* This represents the **minimum cooling capacity** because surroundings and air cooling effects are neglected.

**Step 2: Proposed Method**

The feasibility of the existing cooling unit is evaluated by experimentally determining its actual cooling capacity and comparing it with the required cooling load calculated for the milk storage application. The method is based on an air-side energy balance, supported by measured psychrometric properties and airflow parameters.

**Principle of the Method**<br>
The cooling system removes heat from the cold room through circulating air. As air passes through the evaporator, its enthalpy decreases. Therefore, the rate of heat removal from the room can be determined from:

$$\dot{Q} = \dot{m}_{air} (h_{return} - h_{supply})$$

**Where:**
* $\dot{Q}$ = cooling capacity $(\text{W})$
* $\dot{m}_{air}$ = mass flow rate of air $(\text{kg/s})$
* $h_{return}, h_{supply}$ = enthalpy of return and supply air $(\text{kJ/kg})$
<br>

**Establishing Initial Conditions**<br>
The cold room is first brought to an elevated temperature (approximately $24^\circ\text{C}$) by allowing controlled heat ingress. Once the desired temperature is reached, the room is sealed to begin the cooling experiment.

**Cooling Test**<br>
The refrigeration unit is operated under normal conditions as the room temperature decreases from approximately $24^\circ\text{C}$ to $20^\circ\text{C}$. At 1-minute intervals, the following parameters are recorded:
* Return air dry-bulb temperature ($\text{DBT}$) and wet-bulb temperature ($\text{WBT}$)
* Supply air dry-bulb temperature ($\text{DBT}$) and wet-bulb temperature ($\text{WBT}$)
* Air velocity in the duct ($V$)

**Mass Flow Rate of Air**<br>
The mass flow rate of air is determined using:
$$\dot{m}_{air} = \rho \cdot A \cdot V$$

Where:
* $\rho$ = density of air (evaluated at mean temperature)
* $A$ = cross-sectional area of duct
* $V$ = average air velocity

**Determination of Air Enthalpy**<br>
The enthalpies ($h$) of return and supply air are obtained using standard psychrometric relations based on measured $\text{DBT}$ and $\text{WBT}$ values.

**Cooling Capacity Calculation**<br>
At each time interval, the instantaneous cooling capacity is calculated as:
$$\dot{Q}(t) = \dot{m}_{air} (h_{return} - h_{supply})$$

The average cooling capacity over $n$ readings is determined:
$$\dot{Q}_{avg} = \frac{1}{n} \sum_{i=1}^{n} \dot{Q}(t_i)$$

**Feasibility Assessment**<br>
The experimentally obtained average cooling capacity is compared with the required cooling load ($\dot{Q}_{required} \approx 3.03 \text{ kW}$).

* If $\dot{Q}_{avg} \geq \dot{Q}_{required}$: The system is considered feasible.
* Otherwise: The system is insufficient, and modifications are required.
