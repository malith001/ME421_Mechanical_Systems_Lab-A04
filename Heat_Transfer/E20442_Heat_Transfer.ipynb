## 2. Empirical Assessment of HVAC Cooling Performance


## 2.1 Test Apparatus and Sensors

Airflow properties are measured using a standard duct-traverse technique to fully characterize the air stream's thermo-physical state.

- **Duct Dimensions:** Cross-section area $A = w \times h$  
- **Flow Speed:** Point-wise velocities $v$ obtained with multi-point anemometer grid  
- **Air Conditions:** Dry-bulb $T_{db}$ and wet-bulb $T_{wb}$ temperatures logged at inlet (point 1) and outlet (point 2)  

---

## 2.2 Calculation Procedure

Cooling output is computed from air mass flow and the enthalpy drop across the unit.

### Step 1: Air Mass Flow Rate

$$
\dot{m}_a = \rho \cdot A \cdot \bar{v}
$$

Here, $\rho \approx 1.2 \ \text{kg/m}^3$ is air density and $\bar{v}$ is the average velocity.

---

### Step 2: Actual Cooling Rate

Total cooling (sensible + latent) follows the energy balance:

$$
\dot{Q}_{act} = \dot{m}_a (h_1 - h_2) = \dot{m}_a \Delta h
$$

Enthalpies $h_1$ and $h_2$ are obtained from psychrometric charts using measured $T_{db}$ and $T_{wb}$.

---

## 2.3 Measurement Summary Table

| Parameter          | Symbol             | Unit              | Recorded Value |
|--------------------|-------------------|-------------------|----------------|
| **Duct Area**      | $A$               | $\text{m}^2$      |                |
| **Avg. Velocity**  | $\bar{v}$         | $\text{m/s}$      |                |
| **Inlet Enthalpy** | $h_1$             | $\text{kJ/kg}$    |                |
| **Outlet Enthalpy**| $h_2$             | $\text{kJ/kg}$    |                |
| **Cooling Output** | $\dot{Q}_{act}$   | $\text{kW}$       |                |

---

## 2.4 Acceptance Criteria

Unit performance is validated against the $3.03 \ \text{kW}$ design requirement:

1. **Adequate Capacity:**  
$$
\dot{Q}_{act} \geq 3.03 \ \text{kW}
$$

2. **Sufficient Margin (preferred):**  
$$
\dot{Q}_{act} > 3.03 \ \text{kW}
$$

# 4. Thermodynamic Analysis Framework



This section describes the analytical procedure used to convert raw sensor data into the system’s **Actual Cooling Capacity** $\dot{Q}_{act}$.

---

## 4.1 Air Mass Flow Rate

The mass flow of air is the primary driver of heat transfer. It is obtained from the duct geometry and inlet velocity using the continuity equation:

$$
\dot{m}_a = \rho \cdot A \cdot V_{in}
$$

where:

- $\rho \approx 1.2 \ \text{kg/m}^3$ is air density  
- $A$ is the duct cross-sectional area  
- $V_{in}$ is the measured inlet air velocity  

---

## 4.2 Psychrometric Enthalpy Model

To capture both **sensible** (temperature) and **latent** (moisture) cooling, the specific enthalpy $h$ is derived using the Magnus–Tetens saturation vapor pressure model.

---

### Saturation and Partial Vapor Pressure

First, the saturation vapor pressure at wet-bulb temperature $T_{wb}$ is:

$$
P_{swb} =
0.61078 \cdot
\exp\left(
\frac{17.27 \cdot T_{wb}}{T_{wb} + 237.3}
\right)
$$

Then, the actual vapor partial pressure:

$$
P_v =
P_{swb}
- 0.000662 \cdot P_{atm} \cdot (T_{db} - T_{wb})
$$

---

### Humidity Ratio

$$
\omega =
0.62194 \cdot
\frac{P_v}{P_{atm} - P_v}
$$

---

### Specific Enthalpy

$$
h =
(c_{p,a} \cdot T_{db})
+
\omega \cdot (h_{fg} + c_{p,v} \cdot T_{db})
$$

**Constants:**

- $c_{p,a} = 1.006 \ \text{kJ/kg·K}$  
- $h_{fg} = 2501 \ \text{kJ/kg}$  
- $c_{p,v} = 1.86 \ \text{kJ/kg·K}$  
- $P_{atm} = 101.325 \ \text{kPa}$  

---

## 4.3 System Cooling Capacity

The **Actual Cooling Capacity** is:

$$
\dot{Q}_{act} =
\dot{m}_a \cdot (h_{inlet} - h_{outlet})
$$

where:

- $h_{inlet}$ and $h_{outlet}$ are obtained from measured $T_{db}$ and $T_{wb}$  

---

## 4.4 Performance Index

The effectiveness of each fan-speed setting is evaluated using:

$$
\eta =
\frac{\dot{Q}_{act}}{\dot{Q}_{req}}
$$

where:

$$
\dot{Q}_{req} = 3.03 \ \text{kW}
$$

A value of $\eta \geq 1$ indicates that the system meets the design requirement.

```python
import numpy as np
import pandas as pd

def calculate_enthalpy(Tdb, Twb):
    """
    Calculates specific enthalpy (h) in kJ/kg based on Dry Bulb and Wet Bulb temperatures.
    Uses the formulas established in Section 5.
    """
    Patm = 101.325  # kPa

    # I. Saturation Vapor Pressure at Wet Bulb (Pswb)
    Pswb = 0.61078 * np.exp((17.27 * Twb) / (Twb + 237.3))

    # II. Actual Vapor Pressure (Pv)
    Pv = Pswb - 0.000662 * Patm * (Tdb - Twb)

    # III. Humidity Ratio (w)
    w = 0.62194 * (Pv / (Patm - Pv))

    # IV. Total Specific Enthalpy (h)
    # cp_a = 1.006, hfg = 2501, cp_v = 1.86
    h = (1.006 * Tdb) + w * (2501 + 1.86 * Tdb)

    return h

# Experimental Data
data = {
    'Speed': range(1, 11),
    'T3_db': [29.1, 29.1, 29.5, 29.6, 29.3, 29.3, 29.3, 29.1, 28.9, 29.3],
    'T4_wb': [21.2, 21.1, 21.5, 21.8, 21.2, 21.1, 21.1, 21.0, 20.9, 21.1],
    'T5_db': [26.4, 26.7, 24.3, 20.4, 20.8, 21.4, 21.5, 21.2, 21.3, 19.9],
    'T6_wb': [19.9, 20.0, 19.7, 16.6, 16.2, 18.0, 18.2, 17.6, 17.4, 16.5],
    'Vin':   [0.0, 0.0, 0.6, 2.8, 5.6, 8.8, 9.3, 9.0, 9.8, 9.5]
}

df = pd.DataFrame(data)

# Constants
rho = 1.2  # kg/m3
A = 0.05   # EDIT THIS: Enter your actual Duct Area in m^2

# Process Calculations
df['h_in'] = df.apply(lambda row: calculate_enthalpy(row['T3_db'], row['T4_wb']), axis=1)
df['h_out'] = df.apply(lambda row: calculate_enthalpy(row['T5_db'], row['T6_wb']), axis=1)
df['m_dot'] = rho * A * df['Vin']
df['Q_act_kW'] = df['m_dot'] * (df['h_in'] - df['h_out'])

# Comparison against Design Load (3.03 kW)
df['Requirement_Met'] = df['Q_act_kW'] >= 3.03

# Results
print("--- Evaporator Performance Analysis ---")
display(df[['Speed', 'Vin', 'h_in', 'h_out', 'Q_act_kW', 'Requirement_Met']])
```

# 5. System Scaling for 80-Container Operation



This section shows how the refrigeration system must be adjusted when the daily throughput increases to **80 containers**, while keeping the cooling time fixed at **2 hours (7200 s)**.

---

## 5.1 Updated Cooling Load

When the number of containers doubles, the total heat energy to be removed also doubles within the same time window.

### Total Heat Energy to Remove

$$
Q_{new} = 2 \cdot Q_{max} = 2 \cdot 21{,}842{,}200 \ \text{J}
$$

$$
Q_{new} \approx 43{,}684{,}400 \ \text{J} \approx 43.68 \ \text{MJ}
$$

---

### Required Cooling Power

$$
\dot{Q}_{req,80} = \frac{Q_{new}}{7200}
$$

$$
\dot{Q}_{req,80} \approx 6.06 \ \text{kW}
$$

---
