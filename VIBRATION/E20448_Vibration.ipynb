In flexural vibration, a beam moves in a transverse (lateral) direction, meaning the motion occurs perpendicular to its longitudinal axis. Unlike a vibrating stringwhich relies entirely on external tension to pull it back to centera beam resists deformation through its own internal structural integrity. As it bends, the beams cross-section generates internal bending moments and shear forces, which act as the restoring forces that drive the vibration.

The Bernoulli-Euler theory governs the lateral vibrations of thin beams by modeling their movement in the $x-y$ plane. In this setup, $x$ identifies the position along the beams length, while $y$ measures the vertical bending or deflection at any given point. By examining the forces and moments acting on a microscopic beam element of length $\delta x$, we can formulate a differential equation that balances the beams internal stiffness against its inertial mass.


$$
A \frac{ ^2 v(x,t)}{ t^2}
+ c \frac{  v(x,t)}{ t}
+ P \frac{\partial^2 v(x,t)}{\partial x^2}
+ \frac{^2}{ x^2} \left(EI \frac{^2 v(x,t)}{ x^2}\right)
= f(x,t)
$$

 : mass density of the beam material ($kg/m^3$)

A : cross-sectional area of the beam ( $m^2$)

E : Young's modulus of the beam material ($N/m^2$)

I : second moment of area of the beam cross-section ($m^4$)

c : viscous damping coefficient per unit length ($N \cdot s/m^2$)

v(x,t) : transverse displacement of the beam at position ($x$) and time ( $t$ $m$)

f(x,t) : distributed transverse force per unit length acting on the beam ($N/m$)



$$
\rho A \frac{\partial^2 v(x,t)}{\partial t^2}
+ EI \frac{\partial^4 v(x,t)}{\partial x^4}
+ c\ \delta(x-l_d) \frac{\partial v(x,t)}{\partial t}
=
F_e\sin(_e t)\,\delta(x-l_e)\tag{1}
$$

 : mass density of the beam material ($kg/m^3$)

b : Breadth (m)

h : Thickness (m)

A : cross-sectional area of the beam ($(b.h)$ $m^2$)

E : Young's modulus of the beam material ($N/m^2$)

I : second moment of area of the beam cross-section ($m^4$)

v(x,t) : transverse displacement of the beam at position ($x$) and time ($t$ $m$)

EI : flexural rigidity of the beam ($N\cdot m^2$)

c : viscous damping coefficient of the dashpot ($N\cdot s/m$)

l_d : position of the dashpot measured from the reference end of the beam ($m$)

$F_e$ : amplitude of harmonic point force ($N$)

$_e$ : forcing angular frequency ($rad/s$)

l_e : position of the exciter (point force location) measured from the reference end ($m$)

(x-l) : Dirac delta function (represents a force/damper applied at a single point)

# Modal Analysis



Since the system vibrates in a consistent, fixed geometry for any given mode, we can treat the beam's deflection as a product of two distinct variables: a shape function ($x$) and a time function ($t$)

Assume solution for $v(x,t)$, is in form of,

$$v(x,t) = Y(x) . q(t) \tag{3}$$


$Y(x)$ - Eigen Function/ Mode shape

$q(t)$ - Generalized Corrdinate

Calculating partial derivatives of $v(x,t)$


Velocity
$$\frac{\partial v(x,t)}{\partial t} = Y(x) \dot{q}(t)$$

Acceleration
$$\frac{\partial^2 v(x,t)}{\partial t^2} = Y(x) \ddot{q}(t)$$

Beam Curvature/Flexure
$$\frac{\partial^4 v(x,t)}{\partial x^4} = \frac{d^4 Y(x)}{dx^4} q(t) = Y''''(x) q(t)$$


Substitute into the governing equation (2)

$$\rho A . Y(x) \ddot{q}(t) + EI. Y''''(x) q(t) = 0 \tag{4}$$
