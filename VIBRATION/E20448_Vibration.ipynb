In flexural vibration, a beam moves in a transverse (lateral) direction, meaning the motion occurs perpendicular to its longitudinal axis. Unlike a vibrating string—which relies entirely on external tension to pull it back to center—a beam resists deformation through its own internal structural integrity. As it bends, the beam’s cross-section generates internal bending moments and shear forces, which act as the restoring forces that drive the vibration.

The Bernoulli-Euler theory governs the lateral vibrations of thin beams by modeling their movement in the $x-y$ plane. In this setup, $x$ identifies the position along the beam’s length, while $y$ measures the vertical bending or deflection at any given point. By examining the forces and moments acting on a microscopic beam element of length $\delta x$, we can formulate a differential equation that balances the beam’s internal stiffness against its inertial mass.


$$
ρA \frac{∂ ^2 v(x,t)}{∂ t^2}
+ c \frac{∂  v(x,t)}{∂ t}
+ P \frac{\partial^2 v(x,t)}{\partial x^2}
+ \frac{∂^2}{∂ x^2} \left(EI \frac{∂^2 v(x,t)}{∂ x^2}\right)
= f(x,t)
$$

ρ : mass density of the beam material ($kg/m^3$)

A : cross-sectional area of the beam ( $m^2$)

E : Young's modulus of the beam material ($N/m^2$)

I : second moment of area of the beam cross-section ($m^4$)

c : viscous damping coefficient per unit length ($N \cdot s/m^2$)

v(x,t) : transverse displacement of the beam at position ($x$) and time ( $t$ $m$)

f(x,t) : distributed transverse force per unit length acting on the beam ($N/m$)



$$
\rho A \frac{\partial^2 v(x,t)}{\partial t^2}
+ EI \frac{\partial^4 v(x,t)}{\partial x^4}
+ c\ \delta(x-l_d) \frac{\partial v(x,t)}{\partial t}
=
F_e\sin(ω_e t)\,\delta(x-l_e)\tag{1}
$$

# Separation of Variables: Euler-Bernoulli Beam

## Starting from Undamped Equation
For the **undamped** beam equation (no forcing, no damping):

$$
\rho A \frac{\partial^2 v(x,t)}{\partial t^2} + EI \frac{\partial^4 v(x,t)}{\partial x^4} = 0 \tag{A}
$$

## Assume Product Solution
$$
v(x,t) = Y(x) q(t)
$$

**Substitute:**
$$
\rho A Y(x) \ddot{q}(t) + EI Y''''(x) q(t) = 0 \tag{B}
$$

## Divide Through by ρA·Y(x)·q(t)
$$
\frac{\ddot{q}(t)}{q(t)} + \frac{EI}{\rho A} \frac{Y''''(x)}{Y(x)} = 0 \tag{C}
$$

## Separate Variables
Rearrange to isolate $x$ and $t$ dependencies:

$$
\frac{EI}{\rho A} \frac{Y''''(x)}{Y(x)} = -\frac{\ddot{q}(t)}{q(t)} \tag{D}
$$

**Key insight:** Left = $f(x)$ only, Right = $g(t)$ only → both = **constant**

## Set Separation Constant = ω²
For harmonic motion ($q(t) = Q \sin(\omega t + \phi)$):

$$
\frac{EI}{\rho A} \frac{Y''''(x)}{Y(x)} = -\frac{\ddot{q}(t)}{q(t)} = \omega^2 \tag{5}
$$
