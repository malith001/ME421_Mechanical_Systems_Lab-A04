### Activity #1: Derivation of the 2-DOF Freedom Model

To derive the 2-DOF (two-degree-of-freedom) model, we'll consider the system's components and their equivalent masses, stiffnesses, and damping. The goal is to represent the system's dominant modes of small amplitude vibration using two generalized coordinates.

**1. Identify Degrees of Freedom:**
We assume small amplitude vertical vibrations and focus on the first two dominant bending modes. Therefore, we choose two generalized coordinates:

*   $x_1(t)$: Vertical displacement at the spring end (Point A in the diagram).
*   $x_2(t)$: Vertical displacement at the motor location (Point B in the diagram).

**2. System Components and Equivalent Parameters:**

*   **Rigid Beam:** For small deflections, the beam contributes to both mass and stiffness. Its distributed mass and stiffness are often represented by equivalent lumped masses and springs at the chosen degrees of freedom.

*   **Left End - Pinned (Hinge):** This end has zero vertical displacement, but allows rotation. It acts as a fixed support for the beam's bending behavior. This implies that only the other end of the beam will show vertical displacement when forced.

*   **Right End - Linear Spring:** This spring contributes stiffness directly to $x_1(t)$. The beam itself also has an equivalent stiffness at this point due to its bending properties. Let's denote the stiffness of the external spring as $k_{ext}$.

*   **Middle - Unbalanced Rotating Motor:** This motor acts as an excitation source and contributes mass at $x_2(t)$. The beam's stiffness also plays a role here.


**3. Formulation of the Equations of Motion:**

Using Newton's second law ($F=ma$) or Lagrange's equations, we can write the equations of motion for each degree of freedom. For a 2-DOF system with damping and excitation, the general form is:

$$ M\ddot{X} + C\dot{X} + KX = F(t) $$

Where:
*   **M** is the mass matrix.
*   **C** is the damping matrix.
*   **K** is the stiffness matrix.
*   **X** is the displacement vector $\begin{bmatrix} x_1(t) \ x_2(t) \\ \end{bmatrix}$.
*   **F(t)** is the excitation force vector.

Let's assume the provided scalar equations represent the derived model. The notebook already provides a detailed scalar form, so let's translate that into the matrix form more explicitly.

**Given Scalar Equations (from the notebook):**

$m_1\ddot{x}_1+(c_1+c_2)\dot{x}_1+(k_1+k_2)x_1 -c_2\dot{x}_2-k_2x_2 =f_1(t)$,

$m_2\ddot{x}_2+c_2\dot{x}_2+k_2x_2-c_2\dot{x}_1-k_2x_1 =f_2(t)$,

We can organize these into the standard matrix form $M\ddot{X} + C\dot{X} + KX = F(t)$:

**Mass Matrix (M):**
The terms associated with accelerations ($\ddot{x}_1$ and $\ddot{x}_2$) form the mass matrix. Since $m_1$ is only with $\ddot{x}_1$ and $m_2$ with $\ddot{x}_2$, and no cross-acceleration terms exist, M is a diagonal matrix:

$$ M = \begin{bmatrix} m_1 & 0 \\ 0 & m_2 \end{bmatrix} $$

**Damping Matrix (C):**
The terms associated with velocities ($\dot{x}_1$ and $\dot{x}_2$) form the damping matrix:
*   Coefficient of $\dot{x}_1$ in the first equation: $(c_1+c_2)$
*   Coefficient of $\dot{x}_2$ in the first equation: $-c_2$
*   Coefficient of $\dot{x}_1$ in the second equation: $-c_2$
*   Coefficient of $\dot{x}_2$ in the second equation: $c_2$

$$ C = \begin{bmatrix} (c_1+c_2) & -c_2 \\ -c_2 & c_2 \end{bmatrix} $$

**Stiffness Matrix (K):**
The terms associated with displacements ($x_1$ and $x_2$) form the stiffness matrix:
*   Coefficient of $x_1$ in the first equation: $(k_1+k_2)$
*   Coefficient of $x_2$ in the first equation: $-k_2$
*   Coefficient of $x_1$ in the second equation: $-k_2$
*   Coefficient of $x_2$ in the second equation: $k_2$

$$ K = \begin{bmatrix} (k_1+k_2) & -k_2 \\ -k_2 & k_2 \end{bmatrix} $$

**Force Vector (F(t)):**
The external forces acting on each degree of freedom:

$$ F(t) = \begin{bmatrix} f_1(t) \\ f_2(t) \end{bmatrix} $$

Therefore, the complete 2-DOF model in matrix form is:

$$ \begin{bmatrix} m_1 & 0 \\ 0 & m_2 \end{bmatrix} \begin{bmatrix} \ddot{x}_1 \\ \ddot{x}_2 \end{bmatrix} + \begin{bmatrix} (c_1+c_2) & -c_2 \\ -c_2 & c_2 \end{bmatrix} \begin{bmatrix} \dot{x}_1 \\ \dot{x}_2 \end{bmatrix} + \begin{bmatrix} (k_1+k_2) & -k_2 \\ -k_2 & k_2 \end{bmatrix} \begin{bmatrix} x_1 \\ x_2 \end{bmatrix} = \begin{bmatrix} f_1(t) \\ f_2(t) \end{bmatrix} $$

### Activity #2: Estimate the damping ratios of the first two dominant modes

To estimate the damping ratios for the first two dominant modes of a two-degree-of-freedom (2-DOF) system, we will apply modal decomposition principles. We will calculate the system's natural frequencies and mode shapes, and then use these to determine the modal damping ratios based on a Rayleigh damping assumption.



To estimate the damping ratios for the first two dominant modes of a two-degree-of-freedom (2--DOF) system, modal decomposition principles are applied. The damping ratio for each mode is obtained by transforming the physical damping matrix into modal coordinates using the system eigenvectors (mode shapes).


The equations of motion for a linear 2--DOF system are written as:
\begin{equation}
\mathbf{M}\,\ddot{\mathbf{x}}(t) + \mathbf{C}\,\dot{\mathbf{x}}(t) + \mathbf{K}\,\mathbf{x}(t) = \mathbf{0}
\end{equation}

where:
\begin{align}
\mathbf{M} &=
\begin{bmatrix}
m_1 & 0 \\
0 & m_2
\end{bmatrix},
&
\mathbf{C} &=
\begin{bmatrix}
c_1 + c_2 & -c_2 \\
-c_2 & c_2
\end{bmatrix},
&
\mathbf{K} &=
\begin{bmatrix}
k_1 + k_2 & -k_2 \\
-k_2 & k_2
\end{bmatrix}
\end{align}


The undamped free vibration problem is defined by:
\begin{equation}
\left( \mathbf{K} - \omega^2 \mathbf{M} \right)\boldsymbol{\phi} = \mathbf{0}
\end{equation}

Solving the characteristic equation:
\begin{equation}
\det \left( \mathbf{K} - \omega^2 \mathbf{M} \right) = 0
\end{equation}

yields the natural frequencies:
\begin{equation}
\omega_1,\; \omega_2
\end{equation}


For each natural frequency $\omega_i$, the corresponding mode shape vector $\boldsymbol{\phi}_i$ is obtained from:
\begin{equation}
\left( \mathbf{K} - \omega_i^2 \mathbf{M} \right)\boldsymbol{\phi}_i = \mathbf{0}
\end{equation}

The modal matrix is then defined as:
\begin{equation}
\boldsymbol{\Phi} = \begin{bmatrix} \boldsymbol{\phi}_1 & \boldsymbol{\phi}_2 \end{bmatrix}
\end{equation}

\subsubsection{Modal Parameters}

The modal mass and modal damping for mode $i$ are given by:
\begin{align}
m_i &= \boldsymbol{\phi}_i^{T}\mathbf{M}\boldsymbol{\phi}_i \\
c_i &= \boldsymbol{\phi}_i^{T}\mathbf{C}\boldsymbol{\phi}_i
\end{align}


The damping ratio for each mode is calculated as:
\begin{equation}
\zeta_i = \frac{c_i}{2 m_i \omega_i}
\end{equation}



If Rayleigh damping is assumed:
\begin{equation}
\mathbf{C} = \alpha \mathbf{M} + \beta \mathbf{K}
\end{equation}

the modal damping ratio becomes:
\begin{equation}
\zeta_i = \frac{1}{2}\left( \frac{\alpha}{\omega_i} + \beta \omega_i \right)
\end{equation}


##Activity #3

Plot the following:

* the unforced repsonse to an initiail diplacement of the spring end of the beam
* the forced response for three choices of the rotational speed of the motor
* theoretical frequency response of the system that captures the first two dominant modes of vibration of the system.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# --- 1. System Parameters ---
# These values are illustrative. You can replace them with your calculated
# mass (kg) and stiffness (N/m) values from your specific design.
m1, m2 = 2.0, 1.0
k1, k2 = 2000.0, 1000.0

# Define Mass (M) and Stiffness (K) Matrices
M = np.array([[m1, 0],
              [0, m2]])

K = np.array([[k1 + k2, -k2],
              [-k2, k2]])

# --- 2. Modal Analysis (Natural Frequencies) ---
# Solve the eigenvalue problem: det(K - w^2 M) = 0
evals, evecs = np.linalg.eig(np.linalg.inv(M) @ K)

# Sort eigenvalues to ensure Mode 1 is the lowest frequency
idx = evals.argsort()
evals = evals[idx]
omega_n = np.sqrt(evals) # Natural frequencies in rad/s

print(f"Natural Frequencies: {omega_n} rad/s")

# --- 3. Damping Matrix (Rayleigh Damping) ---
# We estimate damping ratios (zeta) for the two modes (e.g., 5% each)
zeta = np.array([0.05, 0.05])

# Solve for Rayleigh coefficients alpha and beta:
# zeta_i = alpha/(2*omega_i) + beta*omega_i/2
A = np.array([
    [1/(2*omega_n[0]), omega_n[0]/2],
    [1/(2*omega_n[1]), omega_n[1]/2]
])
alpha, beta = np.linalg.solve(A, zeta)

# Construct the Damping Matrix C
C = alpha * M + beta * K

# --- 4. Define the System Dynamics (ODE) ---
# State vector y = [x1, x2, v1, v2] (displacement and velocity)
def system_dynamics(t, y, omega_f=0, F0=0):
    x = y[:2]  # Displacements
    v = y[2:]  # Velocities

    # Forcing function: Applies only to Mass 1 (the "spring end")
    # You can adjust which mass receives the force by changing the array below
    F = np.array([F0 * np.sin(omega_f * t), 0])

    # Equation of Motion: M*accel + C*v + K*x = F
    # accel = M_inv * (F - C*v - K*x)
    accel = np.linalg.inv(M) @ (F - C @ v - K @ x)

    return np.concatenate([v, accel])

# --- 5. Generate Plot Data ---

# A. Unforced Response (Free Vibration)
# Initial condition: Mass 1 displaced by 1cm (0.01m), zero velocity
y0_unforced = [0.01, 0, 0, 0]
t_span = (0, 5) # 5 seconds duration
t_eval = np.linspace(0, 5, 1000)

sol_unforced = solve_ivp(system_dynamics, t_span, y0_unforced, t_eval=t_eval)

# B. Forced Response (Three Rotational Speeds)
# We choose speeds relative to the fundamental natural frequency (omega_n[0])
# 1. Sub-resonant (0.5x), 2. Resonant (1.0x), 3. Intermediate/Super-resonant (1.5x)
omegas_f = [0.5 * omega_n[0], omega_n[0], 1.5 * omega_n[0]]
F0_val = 10.0 # Force amplitude in Newtons
sol_forced = []

for w in omegas_f:
    # Solve for 10 seconds to see steady state
    sol = solve_ivp(system_dynamics, (0, 10), [0, 0, 0, 0],
                    args=(w, F0_val), t_eval=np.linspace(0, 10, 2000))
    sol_forced.append(sol)

# C. Theoretical Frequency Response Function (FRF)
w_range = np.linspace(0, 100, 1000)
mag1 = []
mag2 = []

# Calculate response magnitude |X/F| for a range of frequencies
for w in w_range:
    # Transfer Function Matrix: H(w) = [-w^2 M + i w C + K]^-1
    H_inv = -w**2 * M + 1j * w * C + K
    H = np.linalg.inv(H_inv)

    # Assume unit force applied to Mass 1: F = [1, 0]
    X = H @ np.array([1, 0])

    mag1.append(np.abs(X[0])) # Response of Mass 1
    mag2.append(np.abs(X[1])) # Response of Mass 2

# --- 6. Plotting ---
fig, axs = plt.subplots(3, 1, figsize=(10, 15))

# Plot 1: Unforced
axs[0].plot(sol_unforced.t, sol_unforced.y[0] * 100, label='Mass 1 ($x_1$)')
axs[0].plot(sol_unforced.t, sol_unforced.y[1] * 100, label='Mass 2 ($x_2$)', alpha=0.7)
axs[0].set_title('Unforced Response to Initial Displacement ($x_1(0) = 1$ cm)')
axs[0].set_ylabel('Displacement (cm)')
axs[0].set_xlabel('Time (s)')
axs[0].legend()
axs[0].grid(True)

# Plot 2: Forced
for i, w in enumerate(omegas_f):
    label = f'Speed: {w:.1f} rad/s ({w/(2*np.pi):.1f} Hz)'
    axs[1].plot(sol_forced[i].t, sol_forced[i].y[0] * 100, label=label)
axs[1].set_title('Forced Response for Three Motor Speeds (Mass 1)')
axs[1].set_ylabel('Displacement (cm)')
axs[1].set_xlabel('Time (s)')
axs[1].legend()
axs[1].grid(True)

# Plot 3: Frequency Response
axs[2].semilogy(w_range, mag1, label='Mass 1 ($x_1/F$)')
axs[2].semilogy(w_range, mag2, label='Mass 2 ($x_2/F$)')
axs[2].axvline(omega_n[0], color='r', linestyle='--', label=f'Mode 1: {omega_n[0]:.1f} rad/s')
axs[2].axvline(omega_n[1], color='g', linestyle='--', label=f'Mode 2: {omega_n[1]:.1f} rad/s')
axs[2].set_title('Theoretical Frequency Response (Two Dominant Modes)')
axs[2].set_ylabel('Magnitude (m/N)')
axs[2].set_xlabel('Frequency (rad/s)')
axs[2].legend()
axs[2].grid(True, which='both')

plt.tight_layout()
plt.show()
```
