### Equations of Motion Derivation (Right End Free)

**1. From Newton's Law:**
Assuming $k_3 = 0$ and $c_3 = 0$ from the original system:
$$
\begin{aligned}
m_1\ddot{x}_1 &= -k_1x_1 - c_1\dot{x}_1 + k_2(x_2 - x_1) + c_2(\dot{x}_2 - \dot{x}_1) + F_1 \\
m_2\ddot{x}_2 &= -k_2(x_2 - x_1) - c_2(\dot{x}_2 - \dot{x}_1) + F_2
\end{aligned}
$$

**2. Rearranging terms:**
Grouping variables by displacement, velocity, and acceleration:
$$
\begin{aligned}
m_1\ddot{x}_1 + (c_1 + c_2)\dot{x}_1 + (k_1 + k_2)x_1 - c_2\dot{x}_2 - k_2x_2 &= F_1 \\
m_2\ddot{x}_2 + c_2\dot{x}_2 + k_2x_2 - c_2\dot{x}_1 - k_2x_1 &= F_2
\end{aligned}
$$

**3. Matrix Form:**
The system can be written in the following matrix structure:
$$
\begin{bmatrix} m_1 & 0 \\ 0 & m_2 \end{bmatrix}
\begin{Bmatrix} \ddot{x}_1 \\ \ddot{x}_2 \end{Bmatrix} +
\begin{bmatrix} c_1 + c_2 & -c_2 \\ -c_2 & c_2 \end{bmatrix}
\begin{Bmatrix} \dot{x}_1 \\ \dot{x}_2 \end{Bmatrix} +
\begin{bmatrix} k_1 + k_2 & -k_2 \\ -k_2 & k_2 \end{bmatrix}
\begin{Bmatrix} x_1 \\ x_2 \end{Bmatrix} =
\begin{Bmatrix} F_1 \\ F_2 \end{Bmatrix}
$$

**4. Defining System Matrices:**
$$
\{x\} = \begin{Bmatrix} x_1 \\ x_2 \end{Bmatrix}, \quad \{f\} = \begin{Bmatrix} F_1 \\ F_2 \end{Bmatrix}
$$
$$
[M] = \begin{bmatrix} m_1 & 0 \\ 0 & m_2 \end{bmatrix}, \quad
[C] = \begin{bmatrix} c_1 + c_2 & -c_2 \\ -c_2 & c_2 \end{bmatrix}, \quad
[K] = \begin{bmatrix} k_1 + k_2 & -k_2 \\ -k_2 & k_2 \end{bmatrix}
$$

**Final Form:**
$$[M]\{\ddot{x}\} + [C]\{\dot{x}\} + [K]\{x\} = \{F\}$$
