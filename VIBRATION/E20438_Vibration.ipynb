**The setup in the laboratory; Left end pinned, Excitation at center, right end elasticallly restrtained by a vertical spring.**

It can be assumed that the pinned-pinned axially loaded beam with damping is sufficiently accurately described by the solutions of the partial differential equation given by


$$
\frac{\partial^{2} y}{\partial t^{2}}
+ C\frac{\partial y}{\partial t}+\frac{EI}{\rho A}\frac{\partial^{4} y}{\partial x^{4}}
+ \frac{P}{\rho A}\frac{\partial^{2} y}{\partial x^{2}}
= f(t,,x)
$$



**Assumptions:**

- The beam is slender and follows EulerBernoulli beam theory.
- Transverse deflections are small.
- The material is homogeneous, isotropic, and linearly elastic.
- The axial load $P$ is constant and uniformly distributed.
- Shear deformation and rotary inertia are neglected.
- Damping is negligible (free vibration).


$$
\begin{aligned}
y(x,t) & : \text{Transverse displacement} \\
E & : \text{Young's modulus} \\
I & : \text{Second moment of area} \\
\rho & : \text{Density} \\
A & : \text{Cross-sectional area} \\
P & : \text{Axial load}
\end{aligned}
$$

In this section we will consider the case where the axial load is not present.
So the above equation becomes,

$$
\frac{\partial^{2} y}{\partial t^{2}}
+ C\frac{\partial y}{\partial t}+\frac{EI}{\rho A}\frac{\partial^{4} y}{\partial x^{4}}
= f(t,x) \tag*{-----------------(1)}
$$

According to the apparatus at the laboratory, the boundary conditions become,

**The left end pin supported.** Therefore,

1.No transverse displacement:
$$y(0,t) = 0$$

2.No bending moment:
$$M(0,t) = -EI \frac{\partial^2 y}{\partial x^2}(0,t) = 0 \implies y''(0,t) = 0$$

**The right end is spring supported.** Therefore,

1.No bending moment(Free to rotate):
$$y''(L,t) = 0$$

2.Shearspring equilibrium:

The shear force $V(L,t)$ must be balanced by the spring force $F_s$:
$$V(L,t) = -EI \frac{\partial^3 y}{\partial x^3}(L,t) \quad \text{and} \quad F_s = k_s y(L,t)$$

Force Balance Equation:
$$-EI y'''(L,t) = k_s y(L,t) \implies EI y'''(L,t) + k_s y(L,t) = 0$$

Now we have,

$$\boxed{
\begin{aligned}
x=0: & \quad y(0,t)=0, \quad y''(0,t)=0 \\
x=L: & \quad y''(L,t)=0, \quad EI\,y'''(L,t) + k_s y(L,t) = 0
\end{aligned}
}$$

Let $\mathfrak{F}$ be the infintie dimensional vector space of twice differentiable functions defined on the interval $[0,L]$.
The function space:
$$\mathfrak{F} = \left\{ \phi(x) \in C^{4}[0,L] :
\begin{aligned}
&\phi(0)=0, \quad \phi''(0)=0, \\
&\phi''(L)=0, \quad EI\phi'''(L)+k_s\phi(L)=0
\end{aligned} \right\}$$
This space contains all admissible beam deflection shapes.

Then $H:\mathfrak{F}\mapsto \mathfrak{F}$ defined by
$$H = \frac{EI}{\rho A}\frac{d^4}{dx^4}$$
From equation (1), we can write
$$
\frac{\partial^{2} y}{\partial t^{2}}
+ C\frac{\partial y}{\partial t}+Hy
= f(t,x)
$$

In finite-dimensional vibration theory, the equation of motion of a linear system with damping is:
$$\ddot{x} + c\dot{x} + \Omega x = f(t)$$

where $x(t)$ is a vector of generalized coordinates and $\Omega$ is the stiffness matrix.
