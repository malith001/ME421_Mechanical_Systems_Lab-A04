### Equations of Motion Derivation (Right End Free)

**1. From Newton's Law:**
Assuming $k_3 = 0$ and $c_3 = 0$ from the original system:
$$
\begin{aligned}
m_1\ddot{x}_1 &= -k_1x_1 - c_1\dot{x}_1 + k_2(x_2 - x_1) + c_2(\dot{x}_2 - \dot{x}_1) + F_1 \\
m_2\ddot{x}_2 &= -k_2(x_2 - x_1) - c_2(\dot{x}_2 - \dot{x}_1) + F_2
\end{aligned}
$$

**2. Rearranging terms:**
Grouping variables by displacement, velocity, and acceleration:
$$
\begin{aligned}
m_1\ddot{x}_1 + (c_1 + c_2)\dot{x}_1 + (k_1 + k_2)x_1 - c_2\dot{x}_2 - k_2x_2 &= F_1 \\
m_2\ddot{x}_2 + c_2\dot{x}_2 + k_2x_2 - c_2\dot{x}_1 - k_2x_1 &= F_2
\end{aligned}
$$

**3. Matrix Form:**
The system can be written in the following matrix structure:
$$
\begin{bmatrix} m_1 & 0 \\ 0 & m_2 \end{bmatrix}
\begin{Bmatrix} \ddot{x}_1 \\ \ddot{x}_2 \end{Bmatrix} +
\begin{bmatrix} c_1 + c_2 & -c_2 \\ -c_2 & c_2 \end{bmatrix}
\begin{Bmatrix} \dot{x}_1 \\ \dot{x}_2 \end{Bmatrix} +
\begin{bmatrix} k_1 + k_2 & -k_2 \\ -k_2 & k_2 \end{bmatrix}
\begin{Bmatrix} x_1 \\ x_2 \end{Bmatrix} =
\begin{Bmatrix} F_1 \\ F_2 \end{Bmatrix}
$$

**4. Defining System Matrices:**
$$
\{x\} = \begin{Bmatrix} x_1 \\ x_2 \end{Bmatrix}, \quad \{f\} = \begin{Bmatrix} F_1 \\ F_2 \end{Bmatrix}
$$
$$
[M] = \begin{bmatrix} m_1 & 0 \\ 0 & m_2 \end{bmatrix}, \quad
[C] = \begin{bmatrix} c_1 + c_2 & -c_2 \\ -c_2 & c_2 \end{bmatrix}, \quad
[K] = \begin{bmatrix} k_1 + k_2 & -k_2 \\ -k_2 & k_2 \end{bmatrix}
$$

**Final Form:**
$$[M]\{\ddot{x}\} + [C]\{\dot{x}\} + [K]\{x\} = \{F\}$$

**The setup in the laboratory; Left end pinned, Excitation at center, right end elasticallly restrtained by a vertical spring.**

It can be assumed that the pinned-pinned axially loaded beam with damping is sufficiently accurately described by the solutions of the partial differential equation given by


$$
\frac{\partial^{2} y}{\partial t^{2}}
+ C\frac{\partial y}{\partial x}+\frac{EI}{\rho A}\frac{\partial^{4} y}{\partial x^{4}}
+ \frac{P}{\rho A}\frac{\partial^{2} y}{\partial x^{2}}
= 0
$$



**Assumptions:**

- The beam is slender and follows EulerBernoulli beam theory.
- Transverse deflections are small.
- The material is homogeneous, isotropic, and linearly elastic.
- The axial load $P$ is constant and uniformly distributed.
- Shear deformation and rotary inertia are neglected.
- Damping is negligible (free vibration).


$$
\begin{aligned}
y(x,t) & : \text{Transverse displacement} \\
E & : \text{Young's modulus} \\
I & : \text{Second moment of area} \\
\rho & : \text{Density} \\
A & : \text{Cross-sectional area} \\
P & : \text{Axial load}
\end{aligned}
$$

In this section we will consider the case where the axial load is not present.
So the above equation becomes,

$$
\frac{\partial^{2} y}{\partial t^{2}}
+ C\frac{\partial y}{\partial x}+\frac{EI}{\rho A}\frac{\partial^{4} y}{\partial x^{4}}
= 0
$$
