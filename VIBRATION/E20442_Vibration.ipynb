### **Derivation of a 2-DOF Vibration Model**

By considering a mechanical system undergoing small amplitude vibrations.
\
The system is idealized as a two degree-of-freedom (2-DOF) lumped parameter model.

---

Let:

* ( m_1 , m_2 ) --- lumped masses
* ( k_1 , k_2 ) --- stiffness coefficients
* ( x_1(t), x_2(t) ) --- generalized displacements

Assumptions:

* Linear elastic behavior
* Small displacements
* No damping

---

### **Equations of Motion**

Using **Newtonâ€™s Second Law**, the equations of motion are:

$$[
m_1 \ddot{x}_1 + k_1 x_1 + k_2 (x_1 - x_2) = 0
]$$

$$[
m_2 \ddot{x}_2 + k_2 (x_2 - x_1) = 0
]$$

---

### **Matrix Form**

The equations can be written in matrix form as:

$$[
\mathbf{M}\ddot{\mathbf{x}} + \mathbf{K}\mathbf{x} = \mathbf{0}
]$$

where

$$[
\mathbf{M} =
\begin{bmatrix}
m_1 & 0 \
0 & m_2
\end{bmatrix}
\quad
\mathbf{K} =
\begin{bmatrix}
k_1 + k_2 & -k_2 \
-k_2 & k_2
\end{bmatrix}
]$$

$$[
\mathbf{x} =
\begin{bmatrix}
x_1 \
x_2
\end{bmatrix}
]$$

---

### **Free Vibration Solution**

Assume a harmonic solution:

$$[
\mathbf{x}(t) = \boldsymbol{\phi} e^{i\omega t}
]$$

Substituting into the governing equation gives the **eigenvalue problem**:

$$[
\left( \mathbf{K} - \omega^2 \mathbf{M} \right)\boldsymbol{\phi} = \mathbf{0}
]$$

Non-trivial solutions exist when:

$$[
\det \left( \mathbf{K} - \omega^2 \mathbf{M} \right) = 0
]$$

Solving this equation yields **two natural frequencies**
$(\omega_1)$ and $(\omega_2)$, corresponding to the **first two dominant vibration modes**.

---

### **Conclusion**

The derived **2-DOF lumped parameter model** successfully captures the **first two dominant modes of small-amplitude vibration**.
This model forms the basis for **modal analysis**, **frequency response**, and **dynamic system simulations**.
