### General Equation

When considering the flexural vibration of a beam, the vibration is in the ‚Äútransverse‚Äù or ‚Äúlateral‚Äù direction,
which is accompanied by bending (or flexure) of the member. Hence, the vibrations are perpen-
dicular to the main axis of the member. But a beam, unlike a string can support shear forces and bending moments at
its cross section.

The Bernoulli-Euler equation, which governs transverse vibration of thin beams, consider
a beam in bending, in the x-y plane, with x as the longitudinal axis and y as the transverse axis of
bending deflection.

Considering a small beam element of length Œ¥ùìß we can define the Bernoulli-Euler equation as follow.


$$
œÅA \frac{‚àÇ ^2 v(x,t)}{‚àÇ t^2}
+ c \frac{‚àÇ  v(x,t)}{‚àÇ t}
+ P \frac{\partial^2 v(x,t)}{\partial x^2}
+ \frac{‚àÇ^2}{‚àÇ x^2} \left(EI \frac{‚àÇ^2 v(x,t)}{‚àÇ x^2}\right)
= f(x,t)
$$

œÅ : mass density of the beam material $kg/m^3$

A : cross-sectional area of the beam $m^2$

E : Young's modulus of the beam material $N/m^2$

I : second moment of area of the beam cross-section $m^4$

c : viscous damping coefficient per unit length $N \cdot s/m^2$

v(x,t) : transverse displacement of the beam at position $x$ and time $t$ $m$

f(x,t) : distributed transverse force per unit length acting on the beam $N/m$

## III. Modal Analysis


In each mode, the system will vibrate in a fixed shape ratio. Therefore, the time and space functions

Computing partial derivatives of $v(x,t)$


Velocity (First time derivative):
$$\frac{\partial v(x,t)}{\partial t} = Y(x) \dot{q}(t)$$

Acceleration (Second time derivative):
$$\frac{\partial^2 v(x,t)}{\partial t^2} = Y(x) \ddot{q}(t)$$

### Variable Seperation



To separate the variables, divide the entire equation (4) by $œÅA. Y(x) q(t)$,

$$\frac{\ddot{q}(t)}{q(t)} +  \frac{EI}{\rho A} \frac{Y''''(x)}{Y(x)} = 0$$

$$
\frac{EI}{\rho A} \frac{Y''''(x)}{Y(x)}

## IV. Forced Response

The equation of motion for the beam is given by:
$$
\rho A \frac{\partial^2 v(x,t)}{\partial t^2} + EI \frac{\partial^4 v(x,t)}{\partial x^4} + c_d \delta(x-l_d) \frac{\partial v(x,t)}{\partial t} = F_e \sin(\omega_e t) \delta(x-l_e) \tag{1}
$$

Using the separation of variables for modal expansion:
$$v(x,t) = \sum_{i=1}^{N} Y_i(x) q_i(t) \tag{3}$$

Substituting Equation (3) into (1):
$$
\rho A \sum_{i=1}^{N} Y_i(x) \ddot{q}_i(t) + EI \sum_{i=1}^{N} Y_i''''(x) q_i(t) + c_d \delta(x - l_d) \sum_{i=1}^{N} Y_i(x) \dot{q}_i(t) = F_0 \sin(\Omega t) \delta(x - l_e) \tag{13}
$$

To project onto mode $j$, we multiply Equation (13) by $Y_j(x)$ and integrate from $0$ to $L$:
$$
\sum_{i=1}^{N} \left( \int_{0}^{L} \rho A Y_j(x) Y_i(x) dx \right) \ddot{q}_i(t) + \sum_{i=1}^{N} \left( \int_{0}^{L} EI Y_j(x) Y_i''''(x) dx \right) q_i(t) + c_d \sum_{i=1}^{N} Y_j(l_d) Y_i(l_d) \dot{q}_i(t) = F_0 Y_j(l_e) \sin(\Omega t)
$$
*(Utilizing the property: $\int f(x) \delta(x-a) dx = f(a)$)*

The system can be written in the generalized coupled form:
$$
\sum_{i=1}^{N} M_{ji} \ddot{q}_i(t) + \sum_{i=1}^{N} C_{ji} \dot{q}_i(t) + \sum_{i=1}^{N} K_{ji} q_i(t) = Q_j(t) \tag{14}
$$

Where the modal parameters are defined as:

Mass Matrix: $\displaystyle M_{ji} = \int_{0}^{L} \rho A Y_j(x) Y_i(x) dx$

Damping Matrix: $\displaystyle C_{ji} = c_d Y_j(l_d) Y_i(l_d)$

Stiffness Matrix: $\displaystyle K_{ji} = \int_{0}^{L} EI Y_j(x) Y_i''''(x) dx$

Generalized Force: $\displaystyle Q_j(t) = F_0 Y_j(l_e) \sin(\Omega t)$

#Week 1 ‚Äì Activity 1

A two degree of freedom lumped parameter model is derived to represent the first two dominant vibration modes of the system. The formulation is based on small amplitude vibration assumptions, linear material behavior, and equivalent mass‚Äìstiffness representation of the beam‚Äìspring‚Äìmotor assembly.

To determine the natural frequencies of the beam-spring system, solve for the eigenvalues $\lambda$ using the characteristic equation:

$$EI\lambda^3 \left( \coth(\lambda L) - \cot(\lambda L) \right) = 2k_s$$

First simplify the numerical search by defining a dimensionless parameter $\beta = \lambda L$. Multiplying the equation by $\frac{L^3}{EI}$ yields:

$$\beta^3 \left( \coth(\beta) - \cot(\beta) \right) = \alpha$$

where the stiffness ratio $\alpha$ is defined as:
$$\alpha = \frac{2k_s L^3}{EI}$$

The roots are found by identifying where the function $F(\beta) = 0$,

$$F(\beta) = \beta^3 \left( \coth(\beta) - \cot(\beta) \right) - \alpha$$

Bracketing - Because $\cot(\beta)$ is periodic and contains singularities at $n\pi$, the code searches within intervals $((n-1)\pi, n\pi)$.

Root Finding - The brentq algorithm is used to find the intersection points within these brackets.

The final natural frequencies in Hz are calculated using the beam properties

$$f_n = \frac{\omega_n}{2\pi} = \frac{\lambda_n^2}{2\pi} \sqrt{\frac{EI}{\rho A}}$$

```python
#@title Solve For Eigenvalues Function Definition
def find_betas(alpha, N=2, eps=1e-6):
    """
    Solves the transcendental equation for the dimensionless eigenvalue beta.
    alpha: Dimensionless stiffness ratio (2*k*L^3 / E*I)
    N: Number of modes to find
    eps: Small offset to avoid division by zero at n*pi
    """
    betas = []
    for n in range(1, N+1):
        # Define search boundaries between singularities of cot(beta)
        # cot(beta) is undefined at 0, pi, 2pi, etc.
        a = (n-1)*np.pi + eps
        b = n*np.pi - eps

        # The characteristic function derived from boundary conditions
        # We look for F(beta) = 0
        def F(beta):
            return beta**3 * (coth(beta) - cot(beta)) - alpha

        # Create a fine grid to detect where the function crosses the x-axis
        xs = np.linspace(a, b, 2000)
        Fs = F(xs)

        # Locate indices where the sign of the function flips
        idx = np.where(np.sign(Fs[:-1]) != np.sign(Fs[1:]))[0]

        if len(idx) == 0:
            raise RuntimeError(f"No root bracket found in interval n={n}")

        # Use Brent's method to find the precise root within the detected bracket
        i0 = idx[0]
        root = brentq(F, xs[i0], xs[i0+1])
        betas.append(root)

    return np.array(betas)

# --- Physical Parameter Calculation ---

# Calculate the dimensionless stiffness ratio 'alpha'
alpha = 2 * k_s * L**3 / (E * I)

# Solve for dimensionless eigenvalues (betas) and physical eigenvalues (lambdas)
betas = find_betas(alpha, N)
lambdas = betas / L  # Relationship: beta = lambda * L

# Calculate Angular Frequency (omega)
beam_factor = np.sqrt((E * I) / (rho * A))
omegas = (lambdas**2) * beam_factor

# Convert rad/s to Hz
freqs_hz = omegas / (2 * np.pi)

# Organize results into a structured DataFrame
df = pd.DataFrame({
    'Mode': range(1, len(betas) + 1),
    'Eigenvalues (1/m)': lambdas,
    'Beam Omega (rad/s)': omegas,
    'Beam Frequency (Hz)': freqs_hz
})
```

```python
#@title Modal Frequency Function Definition
def find_system_frequencies(
    df: pd.DataFrame,
    L: float, rho: float, A: float, E: float, I: float,
    point_masses=None,
    k_s: float = 0.0,
    n_int: int = 4000,
) -> pd.DataFrame:
    """
    Adds columns to df:
      - M_beam (kg) : modal mass of beam only
      - M_added (kg): modal added mass from point masses
      - M_total (kg): total modal mass
      - K_beam (N/m): modal stiffness from beam bending
      - K_spring (N/m): modal stiffness from end spring (k_s*Y(L)^2 if k_s>0)
      - K_total (N/m)
      - Omega_new (rad/s)
      - Frequency_new (Hz)

    Take modes Y_j(x) from your earlier expression:
      Y(x)=sin(Œªx)+C*sinh(Œªx), with C = sin(ŒªL)/sinh(ŒªL)
    """
    if point_masses is None:
        point_masses = []

    lambdas = df["Eigenvalues (1/m)"].to_numpy(dtype=float)

    x = np.linspace(0.0, L, n_int)
    dx = x[1] - x[0]

    # allocate arrays
    M_beam = np.zeros_like(lambdas)
    M_added = np.zeros_like(lambdas)
    K_beam = np.zeros_like(lambdas)
    K_spring = np.zeros_like(lambdas)
    omega_new = np.zeros_like(lambdas)

    for i, lam in enumerate(lambdas):
        Yi = _Y(lam, x)
        Yi_pp = _Ypp(lam, x)
        M_beam[i] = rho * A * np.trapezoid(Yi**2, x)
        added = 0.0
        for m_r, x_r in point_masses:
            Yr = _Y(lam, x_r)
            added += m_r * (Yr**2)
        M_added[i] = added

        # modal stiffness (beam bending)
        K_beam[i] = E * I * np.trapezoid(Yi_pp**2, x)

        # modal stiffness (end translational spring)
        if k_s and k_s > 0:
            YL_norm = _Y(lam, L)
            K_spring[i] = k_s * (YL_norm**2)

        K_total = K_beam[i] + K_spring[i]
        M_total = M_beam[i] + M_added[i]
        omega_new[i] = np.sqrt(K_total / M_total)

    out = df.copy()
    out["M_beam (kg)"] = M_beam
    out["M_added (kg)"] = M_added
    out["M_total (kg)"] = M_beam + M_added
    out["K_beam (N/m)"] = K_beam
    out["K_spring (N/m)"] = K_spring
    out["K_total (N/m)"] = out["K_beam (N/m)"] + out["K_spring (N/m)"]
    out["Omega (rad/s)"] = omega_new
    out["Frequency (Hz)"] = omega_new / (2*np.pi)
    return out
```

```python
#@title Find Modal Frequencies
# Define the discrete masses and their positions along the beam
point_masses = [
    (m_d, l_d),          # Mass of the damper hardware at its mounting point
    (m_e, l_e),          # Mass of the exciter at its mounting point
    (m_f, l_s),          # Mass of the spring fixer assembly at the spring location
    (m_s/3, l_s),        # Effective mass of the spring (alpha * m_s, where alpha ‚âà 1/3)
]

# ompute system frequencies considering both the beam and discrete masses
df = find_system_frequencies(
    df=df,
    L=L, rho=rho, A=A, E=E, I=I,
    point_masses=point_masses,
    k_s=k_s,
    n_int=4000           # Number of integration points for accuracy
)

selected_columns = ['Mode', 'Eigenvalues (1/m)', 'Omega (rad/s)', 'Frequency (Hz)']
print("Table 01 : First two modes of the system")
df[selected_columns]
```

#Week 1 ‚Äì Activity 3

The unforced response due to an initial displacement at the spring end of the beam and the forced responses for three different motor rotational speeds are evaluated. Additionally, the theoretical frequency response function capturing the first two dominant modes is obtained.

For an unbalanced exciter motor, the magnitude of the harmonic force is proportional to the square of the angular frequency $œâ_e$

$$F_0 = m_{off}.r_{off}.œâ_e^2$$

```python
#@title Plot Responses Function Definition
def plot_response(df, x_target, f_motor=0.0, x0=1e-3, t_end=10.0, fs=1000, label=None):
    """
    Unified time response at x_target:
      - Free response when f_motor = 0 (because F0 = m_offset*r_offset*omega_e^2 = 0)
      - Forced response when f_motor > 0

    Reuses your globals: L, l_e, zeta1, zeta2, m_offset, r_offset, and _Y().
    Uses first two modes from df.

    Returns
    -------
    t : (Nt,) array
    x_out : (Nt,) array  displacement at x_target
    q_hist : (2,Nt) array modal coords history
    """

    # --- take first two modes (same as your existing functions) ---
    lam = df["Eigenvalues (1/m)"].to_numpy()[:2]
    w   = df["Omega (rad/s)"].to_numpy()[:2]
    M   = df["M_total (kg)"].to_numpy()[:2]
    z   = np.array([zeta1, zeta2], dtype=float)

    # --- forcing setup ---
    omega_e = 2 * np.pi * float(f_motor)
    F0 = m_offset * r_offset * (omega_e**2)  # -> 0 automatically when f_motor=0

    # mode shapes at: target, spring end (IC), force location
    PhiTarget = np.array([_Y(lam[i], x_target) for i in range(2)], dtype=float)
    PhiL      = np.array([_Y(lam[i], L)        for i in range(2)], dtype=float)
    Phif      = np.array([_Y(lam[i], l_e)      for i in range(2)], dtype=float)

    # initial modal coordinates to match x(L,0)=x0, qdot(0)=0
    denom = PhiL @ PhiL
    alpha = x0 / denom if denom != 0 else 0.0
    q0  = alpha * PhiL
    qd0 = np.zeros(2)

    # time grid
    t = np.linspace(0, t_end, int(fs * t_end) + 1)
    dt = t[1] - t[0]

    # ODE: M*qdd + 2*z*w*M*qd + M*w^2*q = Phif*F0*cos(omega_e*t)
    def qdd(tloc, q, qd):
        forcing = Phif * F0 * np.cos(omega_e * tloc)   # vector (2,)
        return (forcing - (2*z*w*M)*qd - (M*(w**2))*q) / M

    def rk4_step(state, ti):
        q  = state[:2]
        qd = state[2:]

        k1_q  = qd
        k1_qd = qdd(ti, q, qd)

        k2_q  = qd + 0.5*dt*k1_qd
        k2_qd = qdd(ti + 0.5*dt, q + 0.5*dt*k1_q, qd + 0.5*dt*k1_qd)

        k3_q  = qd + 0.5*dt*k2_qd
        k3_qd = qdd(ti + 0.5*dt, q + 0.5*dt*k2_q, qd + 0.5*dt*k2_qd)

        k4_q  = qd + dt*k3_qd
        k4_qd = qdd(ti + dt, q + dt*k3_q, qd + dt*k3_qd)

        q_next  = q  + (dt/6.0)*(k1_q  + 2*k2_q  + 2*k3_q  + k4_q)
        qd_next = qd + (dt/6.0)*(k1_qd + 2*k2_qd + 2*k3_qd + k4_qd)
        return np.hstack([q_next, qd_next])

    # integrate
    state = np.hstack([q0, qd0])
    q_hist = np.zeros((2, len(t)))
    for k in range(len(t)):
        q_hist[:, k] = state[:2]
        if k < len(t) - 1:
            state = rk4_step(state, t[k])

    # physical displacement at x_target
    x_out = PhiTarget @ q_hist

    # plot
    plt.figure(figsize=(10, 4))
    plt.plot(t, x_out, label=label if label else (f"free (f=0)" if f_motor == 0 else f"forced f={f_motor:.3g} Hz"))
    plt.axhline(0, color='black', lw=0.5)
    plt.xlabel("Time (s)")
    plt.ylabel(f"Displacement at x={x_target} m [m]")
    plt.title(f"Response at x={x_target} m (F0={F0:.2e} N")
    plt.grid(True, linestyle="--")
    plt.legend()
    plt.show()

    return t, x_out, q_hist
```

```python
#@title Animate Responses Function Definition
def animate_beam_from_qhist(t, x, q_hist, lam, scale=1.0, title="Beam response", skip=10):
    """
    Animate beam displacement y(x,t) = sum_i Y_i(x)*q_i(t)

    Parameters
    ----------
    t : (Nt,) time array
    x : (Nx,) spatial grid along beam
    q_hist : (n_modes, Nt) modal coordinates history
    lam : (n_modes,) lambdas (1/m) for each mode (for _Y)
    scale : multiplies displacement for visibility
    title : plot title
    skip : frame skipping for speed

    Requires: _Y(lam, x) function in your notebook.
    """
    n_modes = q_hist.shape[0]
    Nt = len(t)

    # Precompute mode shapes on grid
    Phi = np.zeros((n_modes, len(x)))
    for i in range(n_modes):
        Phi[i, :] = np.array([_Y(lam[i], xi) for xi in x])

    # Build y(x,t)
    # y_x_t shape: (Nt, Nx)
    y_x_t = (q_hist.T @ Phi) * scale

    fig, ax = plt.subplots(figsize=(8, 3))
    ax.set_xlim(x.min(), x.max())

    max_disp = np.max(np.abs(y_x_t)) * 1.1
    if max_disp == 0:
        max_disp = 1e-6
    ax.set_ylim(-max_disp, max_disp)

    ax.set_xlabel("Position along beam (m)")
    ax.set_ylabel("Displacement (m)")
    ax.set_title(title)
    ax.grid(True, linestyle="--")

    line, = ax.plot([], [], lw=2)
    time_text = ax.text(0.02, 0.90, "", transform=ax.transAxes)

    frames = range(0, Nt, skip)
    interval_ms = int((t[1] - t[0]) * skip * 1000)

    def init():
        line.set_data([], [])
        time_text.set_text("")
        return line, time_text

    def animate(k):
        line.set_data(x, y_x_t[k, :])
        time_text.set_text(f"t = {t[k]:.3f} s")
        return line, time_text

    anim = FuncAnimation(fig, animate, init_func=init, frames=frames,
                         interval=interval_ms, blit=True)

    plt.close(fig)
    display(HTML(anim.to_html5_video()))
    return anim
```

```python
#@title Response Evaluvator Function Definition
def animate_beam(df, x_target=L, f_motor=6.75, x0=0.0, t_end=6, fs=1000):
  # spatial grid
  x = np.linspace(0, L, 60)

  # get a response first (your existing function or mine)
  t, x_out, q_hist = plot_response(df, x_target, f_motor, x0, t_end, fs)

  # lambdas (first two)
  lam = df["Eigenvalues (1/m)"].to_numpy()[:2]

  # animate
  animate_beam_from_qhist(
      t=t, x=x, q_hist=q_hist, lam=lam,
      scale=1.0,
      title="",
      skip=10
  )
```

## 3. Frequency Response

The theoretical frequency response function was obtained using the $e^{jœât}
 harmonic convention, in which a negative phase indicates a lag of the displacement response relative to the excitation force.

```python
#@title Frequency Response Function
def plot_frf_bode_beam(df, x_obs, f_range_hz,
                            n_modes=2, unwrap_phase=True):
    """
    Bode-style FRF for beam displacement, with frequency axis in Hz.

    Plots:
      - Magnitude: 20*log10(|Y|)
      - Phase: arg(Y)

    Parameters
    ----------
    df : DataFrame
        Contains modal data:
        'Eigenvalues (1/m)', 'Omega (rad/s)', 'M_total (kg)'
    x_obs : float
        Observation position along beam (m)
    f_range_hz : array_like
        Excitation frequencies (Hz)
    n_modes : int
        Number of modes (default = 2)
    unwrap_phase : bool
        Unwrap phase for smooth plot
    """

    x_force = l_e

    # modal data
    lam = df["Eigenvalues (1/m)"].to_numpy()[:n_modes]
    w_r = df["Omega (rad/s)"].to_numpy()[:n_modes]
    M_r = df["M_total (kg)"].to_numpy()[:n_modes]
    zeta = np.array([zeta1, zeta2])[:n_modes]

    f_range_hz = np.asarray(f_range_hz)
    Omega = 2*np.pi*f_range_hz  # convert Hz -> rad/s

    # mode shapes
    Phi_obs = np.array([_Y(lam[r], x_obs) for r in range(n_modes)])
    Phi_f   = np.array([_Y(lam[r], x_force) for r in range(n_modes)])

    # frequency dependent forcing (your exciter model)
    F0 = 2*m_offset * r_offset * Omega**2

    # FRF computation
    H = np.zeros_like(Omega, dtype=complex)

    for r in range(n_modes):
        denom = (w_r[r]**2 - Omega**2
                 + 2j * zeta[r] * w_r[r] * Omega)
        H += (Phi_obs[r] * Phi_f[r] / M_r[r]) * F0 / denom

    # magnitude and phase
    mag_db = 20 * np.log10(1000*np.abs(H))
    phase = np.angle(H)
    if unwrap_phase:
        phase = np.unwrap(phase)

    # ================== plotting ==================
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6), sharex=True)

    ax1.plot(f_range_hz, mag_db, lw=2)
    ax1.set_ylabel("Magnitude (dB)")
    ax1.set_title(f"Bode FRF at x = {x_obs:.3f} m")
    ax1.grid(True, linestyle="--")

    ax2.plot(f_range_hz, phase, lw=2)
    ax2.set_xlabel("Excitation frequency f (Hz)")
    ax2.set_ylabel("Phase (rad)")
    ax2.grid(True, linestyle="--")

    plt.tight_layout()
    plt.show()

    return f_range_hz, H
```

```python
#@title Experimental FRF Bode plot Function
def exp_frf_from_measurements(data, use_peak_amplitude=True, ref_amp=1.0):
    """
    Build experimental FRF-style Bode plot from measured data.

    Input columns (required):
      - f_hz   : Frequency (Hz)
      - p2p_mm : Peak-to-peak amplitude (mm)
      - phase_deg : Phase (deg) where you already set 5 Hz = 0 deg

    Parameters
    ----------
    data : dict OR list of dicts OR (freq_list, p2p_list, phase_list)
        Examples:
          data = {"f_hz":[...], "p2p_mm":[...], "phase_deg":[...]}
          data = [{"f_hz":5.0,"p2p_mm":1.2,"phase_deg":0.0}, ...]
          data = (freq_list, p2p_list, phase_list)
    use_peak_amplitude : bool
        If True, converts peak-to-peak to peak amplitude: A_peak = A_p2p/2
        (This is common for FRF magnitude.)
        If False, uses peak-to-peak directly.
    ref_amp : float
        Reference amplitude for dB scaling (same units as used for magnitude after conversion).
        Default 1.0 (so dB is relative to 1 mm if you keep mm, or 1 m if you convert to m).
        You can set ref_amp=1.0 to just get "20log10(amplitude)" style.

    Returns
    -------
    df_obs : DataFrame
        Observation/raw table
    df_proc : DataFrame
        Processed table with magnitude(dB) and phase(rad)
    """

    # --------- Build observation table ---------
    if isinstance(data, dict):
        df_obs = pd.DataFrame(data)
    elif isinstance(data, list):
        df_obs = pd.DataFrame(data)
    elif isinstance(data, tuple) and len(data) == 3:
        df_obs = pd.DataFrame({
            "f_hz": data[0],
            "p2p_mm": data[1],
            "phase_deg": data[2]
        })
    else:
        raise ValueError("data must be a dict, list of dicts, or a tuple (freq_list, p2p_list, phase_list).")

    required_cols = {"f_hz", "p2p_mm", "phase_deg"}
    if not required_cols.issubset(df_obs.columns):
        raise ValueError(f"Missing columns. Required: {required_cols}")

    # sort by frequency
    df_obs = df_obs.sort_values("f_hz").reset_index(drop=True)

    # print("\nTable 03: Observation Data")
    # display(df_obs)  # works in Colab/Jupyter; if not, replace with print(df_obs)

    # --------- Process data ---------
    f_hz = df_obs["f_hz"].to_numpy(dtype=float)
    p2p_mm = df_obs["p2p_mm"].to_numpy(dtype=float)
    phase_deg = df_obs["phase_deg"].to_numpy(dtype=float)

    # amplitude used for magnitude
    if use_peak_amplitude:
        amp_mm = p2p_mm / 2.0
        amp_label = "A_peak_mm"
    else:
        amp_mm = p2p_mm
        amp_label = "A_p2p_mm"

    # Avoid log10(0)
    amp_mm_safe = np.maximum(amp_mm, 1e-12)

    mag_db = 20.0 * np.log10(amp_mm_safe / ref_amp)
    phase_rad = np.deg2rad(phase_deg)

    df_proc = df_obs.copy()
    df_proc[amp_label] = amp_mm
    df_proc["Mag_dB"] = mag_db
    df_proc["Phase_rad"] = phase_rad

    # print("\nTable 04: Processed Data")
    # display(df_proc)  # works in Colab/Jupyter; if not, replace with print(df_proc)

    # --------- Plot Bode ---------
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6), sharex=True)

    ax1.plot(f_hz, mag_db, marker="o", lw=2)
    ax1.set_ylabel("Magnitude (dB)")
    ax1.grid(True, linestyle="--")

    ax2.plot(f_hz, phase_rad, marker="o", lw=2)
    ax2.set_xlabel("Frequency f (Hz)")
    ax2.set_ylabel("Phase (rad)")
    ax2.grid(True, linestyle="--")

    plt.tight_layout()
    plt.show()

    return df_obs, df_proc

def show_frf_table_figure(df, title="Table"):
    fig, ax = plt.subplots(figsize=(8, 0.5 + 0.35 * len(df)))
    ax.axis("off")

    table = ax.table(
        cellText=df.round(3).values,
        colLabels=df.columns,
        loc="center",
        cellLoc="center"
    )

    table.auto_set_font_size(False)
    table.set_fontsize(9)
    table.scale(1, 1.4)

    ax.set_title(title, fontsize=12, fontweight="bold", pad=10)
    plt.tight_layout()
    plt.show()
```

```python
# @title Extract Damped Frequency Function
def extract_damped_peak(df_or_data, freq_col="f_hz", amp_col="p2p_mm",
                        method="quadratic", return_in_hz=True):
    """
    Extract damped peak frequency and maximum amplitude from experimental FRF data.

    Parameters
    ----------
    df_or_data : DataFrame OR dict
        Must contain frequency and amplitude columns.
        Example dict: {"f_hz":[...], "p2p_mm":[...], "phase_deg":[...]}
    freq_col : str
        Frequency column name (Hz)
    amp_col : str
        Amplitude column name (peak-to-peak in mm, or any amplitude you stored)
    method : str
        "discrete"   -> take max measured point
        "quadratic"  -> 3-point parabola fit around max (better peak estimate)
    return_in_hz : bool
        If True returns peak frequency in Hz; also returns Omega_d (rad/s)

    Returns
    -------
    results : dict
        {
          "f_d_hz": ...,
          "Omega_d_rad_s": ...,
          "A_max": ...,
          "peak_index": ...,
          "method": ...
        }
    """

    # ---- load dataframe ----
    if isinstance(df_or_data, pd.DataFrame):
        df = df_or_data.copy()
    elif isinstance(df_or_data, dict):
        df = pd.DataFrame(df_or_data)
    else:
        raise ValueError("df_or_data must be a pandas DataFrame or a dict.")

    # ---- sort ----
    df = df.sort_values(freq_col).reset_index(drop=True)
    f = df[freq_col].to_numpy(dtype=float)
    A = df[amp_col].to_numpy(dtype=float)

    # ---- discrete peak ----
    i = int(np.nanargmax(A))
    f_peak = f[i]
    A_peak = A[i]

    # ---- quadratic refinement (3-point parabola) ----
    if method.lower() == "quadratic":
        # need neighbors
        if i == 0 or i == len(A) - 1:
            # can't refine at boundaries -> fall back
            method_used = "discrete (peak at boundary)"
        else:
            # Fit parabola y = a x^2 + b x + c through 3 points
            x1, x2, x3 = f[i-1], f[i], f[i+1]
            y1, y2, y3 = A[i-1], A[i], A[i+1]

            M = np.array([
                [x1**2, x1, 1.0],
                [x2**2, x2, 1.0],
                [x3**2, x3, 1.0]
            ], dtype=float)

            a, b, c = np.linalg.solve(M, np.array([y1, y2, y3], dtype=float))

            # vertex x* = -b/(2a)
            if np.isclose(a, 0.0):
                method_used = "discrete (parabola nearly linear)"
            else:
                x_star = -b / (2*a)

                # only accept if vertex is inside neighbor interval
                if (min(x1, x3) <= x_star <= max(x1, x3)):
                    f_peak = float(x_star)
                    A_peak = float(a*x_star**2 + b*x_star + c)
                    method_used = "quadratic"
                else:
                    method_used = "discrete (vertex outside neighbors)"

    else:
        method_used = "discrete"

    Omega_d = 2*np.pi*f_peak

    print("\n" + "="*50)
    print(" DAMPED PEAK RESPONSE IDENTIFICATION")
    print("="*50)
    print(f" Peak frequency (f_d)     : {f_peak:8.3f}  Hz")
    print(f" Damped angular freq (Œ©d): {Omega_d:8.3f}  rad/s")
    print(f" Maximum amplitude       : {A_peak:8.3f}  mm (peak-to-peak)")
    # print(f" Extraction method       : {method_used}")
    # print(f" Peak data index         : {i}")
    print("="*50 + "\n")

    return {
        "f_d_hz": float(f_peak),
        "Omega_d_rad_s": float(Omega_d),
        "A_max": float(A_peak),
        "peak_index": i,
        "method": method_used
    }
```

The apparatus comes with a passive undamed vibration absorber as shown in figure above.

The effective absorber parameters for **one side** are,

Cantilever stiffness,
$$ k_{absside} = \frac{3E_{abs}I_{abs}}{l_{abs}^3}\tag{21}$$

Effective mass,
$$ m_{absside} = m_{absmass} + \frac{33}{140} m_{absbeem}l_{abs}\tag{22}$$

Absorber Frequency,
$$ f_{abs} = \frac{1}{2œÄ} \sqrt{\frac{3E_{abs}I_{abs}}{m_{absside} l_{abs}^3}}\tag{23}$$









When designing the absorber, frequency of absorber is tuned near the operating frequency. In this case the first mode.

$$f_{abs} = f_1$$



```python
#@title Define Absorber Parameters

# --- Absorber Properties (from image data) ------------------------------------
E_a          = 2e+11              # [Pa]     Modulus of Elasticity
I_a           = 7.2179e-12        # [m^4] Moment of inertia for one side
m_a_mass      = 0.121             # [kg] Each moveable mass (m_mass)
m_a_beam      = 0.126             # [kg/m] Beam mass (each side)
m_a_total     = 0.438             # [kg] Total mass of absorber assembly
l_a           = l_e               # [m] Location of absorber connected to main system
k_a           = 1.09e3            # [N/m] Stiffness of beam
```

```python
# @title Calculate Absorber Length
def find_l_abs(f_target_hz, E_abs, I_abs, m_abs_mass, m_abs_beam_per_m):
    """
    Solve absorber length l_abs for:
      f = (1/(2œÄ)) * sqrt( 3 E I / ( (m_mass + (33/140)*m_beam_per_m*l) * l^3 ) )

    Parameters
    ----------
    f_target_hz : float
        Target absorber frequency [Hz]
    E_abs : float
        Young's modulus [Pa]
    I_abs : float
        Second moment of area [m^4]
    m_abs_mass : float
        Tip mass on absorber [kg]
    m_abs_beam_per_m : float
        Absorber beam mass per unit length [kg/m]  (IMPORTANT!)

    Returns
    -------
    l_abs : float
        Required absorber length [m]
    """

    alpha = (33.0 / 140.0) * m_abs_beam_per_m  # [kg/m]
    C = (3.0 * E_abs * I_abs) / ((2.0 * math.pi * f_target_hz) ** 2)  # [kg*m^3] effectively

    # Solve: alpha*l^4 + m_mass*l^3 - C = 0
    coeffs = [alpha, m_abs_mass, 0.0, 0.0, -C]  # l^4, l^3, l^2, l^1, l^0

    roots = np.roots(coeffs)

    # Keep only real, positive roots
    real_pos = [r.real for r in roots if abs(r.imag) < 1e-8 and r.real > 0]

    if not real_pos:
        raise ValueError(f"No positive real solution found. Roots were: {roots}")

    # Usually only one meaningful positive root
    l_abs = min(real_pos)  # pick smallest positive real root (physically reasonable)

    return float(l_abs)

l_abs = find_l_abs(6.42, E_a, I_a, m_a_mass, m_a_beam)  # m_a_beam MUST be kg/m
print("============================================================")
print(f"Required Absorber Length: {l_abs:.4f} m")
print("============================================================")
```

## Modeling New System

Let the displacement of mass at the tip of absober be $z(t)$, and mass and astiffness of absorber be $m_a$ and $k_a$.
Take absorber is connected at $l_a$ in main system.

Then displacement of absorber attachment is,
$$v(l_a,t) = v_a(t) = Y_a q(t)$$
Absorber Force,
$$F_a(t) = k_a(z(t) - Y_a q(t)) \tag{24}$$

Considering the first mode of main system,

$$
M_1 \ddot{q}(t) + C_1 \dot{q}(t) + K_1 q(t) = Q_1(t) \tag{25}
$$

Where the modal parameters are defined as:

$M_1 = \int_0^L  \rho A Y_1^2(x) +  (\text{lumped masses})$

$C_{1} = c_d Y_1^2(l_d)$

$K_{1} = œâ_1^2 M_1$

$Q_i(t)= F_0 Y_1(l_e)\sin(\Omega t)$

By adding the effect of absorber force to equation (25),
$$
M_1 \ddot{q}(t) + C_1 \dot{q}(t) + K_1 q(t) = Q_1(t) -  (k_a(z(t) - Y_a q(t)))Y_a
$$

\
$$
M_1 \ddot{q}(t) + C_1 \dot{q}(t) + (K_1 + k_a Y_a^2) q(t) - k_a Y_a z(t)= Q_1(t) \tag{26}
$$

Considering the mass of the absorber,

$$m_a \ddot{z} + k_a z -k_a Y_a q = 0 \tag{27}$$





Considering on the steady state response,Assume the solution of the $q(t)$ and $z(t)$ are harmonic,

$$q(t) = Q e^{iŒ©t}$$
$$z(t) = Z e^{iŒ©t}$$
$$Q_1(t) = F_0Y_1(l_e) e^{iŒ©t}$$

Then substitute that to equation (26) and (27),

$$ (-\Omega^2 M_1 + i\Omega C_1 + K_1 + k_a Y_a^2)Q - k_a Y_a Z = \hat{Q}_1 \tag{28} $$

$$ -k_a Y_a Q + (k_a - \Omega^2 m_a)Z = 0 \tag{29} $$

From (29),
$$ Z =\frac{k_a Y_a}{k_a - Œ©^2 m_a} Q $$

Substituting to (28),

$$ (-\Omega^2 M_1 + i\Omega C_1 + K_1 + k_a Y_a^2)Q - k_a Y_a Z = \hat{Q}_1  $$


$$\left[ -\Omega^2 M_1 + i\Omega C_1 + K_1 + k_a Y_a^2 - \frac{k_a^2 Y_a^2}{k_a - \Omega^2 m_a} \right] Q = F_0Y_1(l_e) $$


$$Q(\Omega) = \frac{F_0Y_1(l_e)}{(-\Omega^2 M_1 + i\Omega C_1 + K_1 + k_a Y_a^2) - \frac{k_a^2 Y_a^2}{k_a - \Omega^2 m_a}} \tag{30}$$

And absorber tip response:

$$Z(\Omega) = \frac{k_a Y_a}{k_a - \Omega^2 m_a} Q(\Omega) \tag{31}$$




Therefore the verticale displacement for new system ,

$$v_{ss}(x,t) = Y_1(x) . q(t)$$
Where,
$$ Y_1(x) = sin(Œª_1x) + \frac{sin(Œª_1L)}{sinh(Œª_1L)} sinh(Œª_1x)$$

and

$$
q(t) =Q(Œ©)e^{j Œ© t}$$

$$Q(\Omega) = \frac{F_0Y_1(l_e)}{(-\Omega^2 M_1 + i\Omega C_1 + K_1 + k_a Y_a^2) - \frac{k_a^2 Y_a^2}{k_a - \Omega^2 m_a}} \tag{30}$$

```python
#@title Frequency Response Function for Beam + Undamped Absorber
def plot_frf_bode_beam_absorber(df, x_obs, f_range_hz,
                               m_abs, k_abs, x_abs=None,
                               n_modes=2, unwrap_phase=True,
                               return_receptance=False):
    """
    Beam FRF with an *undamped* tuned absorber (m_abs, k_abs) attached at x_abs.

    return_receptance=False: returns displacement response including your exciter model F0(Œ©)
    return_receptance=True : returns receptance FRF = y/F (per unit force), i.e. F0 removed
    """

    x_force = l_e
    if x_abs is None:
        x_abs = x_force

    # modal data
    lam = df["Eigenvalues (1/m)"].to_numpy()[:n_modes]
    w_r = df["Omega (rad/s)"].to_numpy()[:n_modes]
    # M_r = df["M_total (kg)"].to_numpy()[:n_modes]
    # Add lumped mass of absorber assembly
    M_r = df["M_total (kg)"].to_numpy()[:n_modes] + m_a_total * np.array([_Y(lam[r], l_a)**2 for r in range(n_modes)])

    zeta = np.array([zeta1, zeta2])[:n_modes]

    # modal matrices (diagonal)
    K_r = M_r * (w_r**2)
    C_r = 2.0 * zeta * w_r * M_r

    f_range_hz = np.asarray(f_range_hz)
    Omega = 2*np.pi*f_range_hz

    # mode shapes
    Phi_obs = np.array([_Y(lam[r], x_obs)   for r in range(n_modes)])
    Phi_f   = np.array([_Y(lam[r], x_force) for r in range(n_modes)])
    Phi_a   = np.array([_Y(lam[r], x_abs)   for r in range(n_modes)])

    # your exciter force amplitude model
    F0 = m_offset * r_offset * Omega**2

    H = np.zeros_like(Omega, dtype=complex)

    for i, Om in enumerate(Omega):

        # Beam dynamic stiffness (modal)
        Zb = np.diag(K_r - (Om**2)*M_r + 1j*Om*C_r)

        # coupling objects
        Kab = k_abs * np.outer(Phi_a, Phi_a)  # (n,n)
        g   = k_abs * Phi_a                   # (n,)

        # Assemble A [q; Y_a] = b
        A = np.zeros((n_modes + 1, n_modes + 1), dtype=complex)

        A[:n_modes, :n_modes] = Zb + Kab     # <-- plus
        A[:n_modes,  n_modes] = -g           # <-- minus
        # absorber row stays the same:
        A[n_modes, :n_modes]  = -g.T
        A[n_modes,  n_modes]  = (k_abs - (Om**2)*m_abs)


        # RHS
        b = np.zeros(n_modes + 1, dtype=complex)

        if return_receptance:
            # per unit force at x_force
            b[:n_modes] = Phi_f * 1.0
        else:
            b[:n_modes] = Phi_f * F0[i]

        sol = np.linalg.solve(A, b)
        q = sol[:n_modes]

        H[i] = np.dot(Phi_obs, q)



    mag_floor = 10**(-100/20)
    mag_db = 20*np.log10(np.maximum(1000*np.abs(H), mag_floor))

    phase = np.angle(H)
    if unwrap_phase:
        phase = np.unwrap(phase)

    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6), sharex=True)
    ax1.plot(f_range_hz, mag_db, lw=2)
    ax1.set_ylabel("Magnitude (dB)")
    ax1.set_title(f"Beam + undamped absorber FRF at x={x_obs:.3f} m (x_abs={x_abs:.3f} m)")
    ax1.grid(True, linestyle="--")

    ax2.plot(f_range_hz, phase, lw=2)
    ax2.set_xlabel("Excitation frequency f (Hz)")
    ax2.set_ylabel("Phase (rad)")
    ax2.grid(True, linestyle="--")

    plt.tight_layout()
    plt.show()

    return f_range_hz, H
```
